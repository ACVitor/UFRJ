# Algebra vetorial e tensorial

#  Pontos, vetores, e tensores

*   **Pontos**:  elementos do Espaço Euclideano Tridimensional  ${\cal E}$

$${\bf x}, \, {\bf y}, \,{\bf p}, \, \dots, \in {\cal E}$$

*   **Vetores**: elementos do Espaço Vetorial  ${\cal V}$

$${\bf u}, \, {\bf v}, \,{\bf n}, \, \dots, \in {\cal V}$$

* Observação 1: ***Um vetor é a diferença entre dois pontos***:   $$\boxed{{\bf u} = {\bf y} - {\bf x}}$$

* Observação 2: ***A soma de um ponto com um  vetor é ponto***:   $$\boxed{{\bf y} = {\bf x} + {\bf u}}$$

* Observação 3: ***Não existe a soma entre dois pontos ***

* **Tensor**:  *transformação linear* de vetores em vetores
$${\bf A}, {\bf E}, \, {\bf T}, \, \dots, \in {\mbox{Lin}}$$


* Para quaisquer vetores ${\bf u}$ e ${\bf v}$ e qualquer escalar $\alpha\in\mathbb{R}$:

$$
\left\{
\begin{array}{lcr}
{\bf{A}}({\bf u} + {\bf v}) = {\bf{A}}({\bf u}) + {\bf{A}}({\bf v}) \\[5pt]
{\bf{A}}(\alpha{\bf u}) = \alpha {\bf{A}}({\bf u})
\end{array}
\right.
$$



#   Produto interno entre dois vetores, norma ou tamanho de um vetor, base ortonormal

* **Produto interno entre os vetores ${\bf u}$ and ${\bf v}$**:  

$${\bf u}\cdot {\bf v}$$

* **Norma ou tamanho  do vetor ${\bf u}$**:

$$
|{\bf u}| = ({\bf u}\cdot {\bf u})^{\frac{1}{2}}, \quad \mathbf{u}^2 = \mathbf{u}\cdot\mathbf{u}
$$

* Vetor unitário

$$
|{\bf u}|=1
$$

* **Vetores ortogonais ou perpendiculares**

$$
{\bf u}\cdot {\bf v} = 0
$$

* Conjunto de todos vetores perpendiculares ao vetor $\mathbf{v}$:

$$
\{\mathbf{v}\}^\perp = \{\mathbf{u}\in {\cal V}\;\mbox{tal que}\; \mathbf{u}\cdot\mathbf{v}=0\}
$$

* **Base ortonormal**  $\{\mathbf{e}_1,\mathbf{e}_2, \mathbf{e}_3\} $:

$$
\left\{
\begin{array}{lcr}
|\mathbf{e}_1|=|\mathbf{e}_2|=|\mathbf{e}_3|=1 \\[5pt]
\mathbf{e}_1\cdot \mathbf{e}_2 = \mathbf{e}_1\cdot \mathbf{e}_3 = \mathbf{e}_2\cdot \mathbf{e}_3 = 0
\end{array}
\right.
$$

* *Forma mais compacta*: Para $i,j=1, 2, 3$:

$$
\mathbf{e}_i \cdot \mathbf{e}_j = \delta_{ij} =
\left\{
\begin{array}{lcr}
1,\;\mbox{se}\;  i=j \\[5pt]
0,\;\mbox{se}\; i\neq j.
\end{array}
\right.
$$

$\delta_{ij}$: delta de Kronecker




#Componentes de um vetor  na base ortonormal $\{\mathbf{e}_1, \mathbf{e}_2, \mathbf{e}_3\}$

Um vetor $\mathbf{u}$  qualquer  pode ser escrito como a combinação linear  dos vetores da base:
$$
\mathbf{u} =  u_1\mathbf{e}_1 + u_2 \mathbf{e}_2 + u_3 \mathbf{e}_3 = \sum_{i=1}^3 u_i \mathbf{e}_i
$$


* Componentes de $\mathbf{u}$ na base  $\{\mathbf{e}_1, \mathbf{e}_2, \mathbf{e}_3\}$:
$$
u_i = \mathbf{u}\cdot\mathbf{e}_i, \; i=1, 2, 3
$$

* Representação matricial de $\mathbf{u}$:

$$
[\mathbf{u}] =
\left[
\begin{array}{lcr}
u_1 \cr
u_2 \cr
u_3
\end{array}
\right]
$$

* Produto interno de dois vetores

$$
\mathbf{u}\cdot\mathbf{v}= \sum_{i=1}^3 u_i v_i
$$

* Tamanho de um vetor

$$
|\mathbf{u}|^2 = \sum_{i=1}^3 u_i^2
$$

## Componentes de um tensor na base $\{\mathbf{e}_1, \mathbf{e}_2, \mathbf{e}_3\}$


Por definição, um tensor $\mathbf{A}$ qualquer  opera no vetor $\mathbf{u}$ da seguinte forma:

$$
\mathbf{A}(\mathbf{u}) = \mathbf{A}(u_1\mathbf{e}_1 + u_2 \mathbf{e}_2 + u_3 \mathbf{e}_3) = u_1 \mathbf{A}(\mathbf{e}_1) +  u_2 \mathbf{A}(\mathbf{e}_2) +  u_3 \mathbf{A}(\mathbf{e}_3).
$$

Como  $\mathbf{A}(\mathbf{e}_1)$, $\mathbf{A}(\mathbf{e}_2)$, e $\mathbf{A}(\mathbf{e}_3)$ são vetores:

$$
\left\{
\begin{array}{lcr}
\mathbf{A}(\mathbf{e}_1) =  A_{11} \mathbf{e}_1 +  A_{21} \mathbf{e}_2 + A_{31}  \mathbf{e}_3, \cr
\mathbf{A}(\mathbf{e}_2) =  A_{12} \mathbf{e}_1 +  A_{22} \mathbf{e}_2 + A_{32}  \mathbf{e}_3, \cr
\mathbf{A}(\mathbf{e}_3) =  A_{13} \mathbf{e}_1 +  A_{23} \mathbf{e}_2 + A_{32}  \mathbf{e}_3,
\end{array}
\right.
\quad\mbox{ou de forma compacta:}\quad \mathbf{A}(\mathbf{e}_i) = \sum_{j=1}^3 A_{ji} \mathbf{e}_j, \quad i=1, 2, 3
$$


* Componentes de $\mathbf{A}$ na base  na base ortonormal $\{\mathbf{e}_1, \mathbf{e}_2, \mathbf{e}_3\}$

$$
A_{ij} = \mathbf{A}(\mathbf{e}_j)\cdot \mathbf{e}_i
$$

* Representação matricial de $\mathbf{A}$:

$$
[\mathbf{A}] =
\left[
\begin{array}{lcr}
A_{11} \; A_{12} \; A_{13}  \cr
A_{21} \; A_{22} \; A_{23}  \cr
A_{31}  \; A_{32} \; A_{33}
\end{array}
\right]
$$

#Operações com tensores

## 1) A soma  entre os tensores $\mathbf{A}$ e $\mathbf{B}$ é o tensor $\mathbf{A} + \mathbf{B}$:

* Definição:

$$
(\mathbf{A} + \mathbf{B})\mathbf{u} = \mathbf{A}\mathbf{u} +  \mathbf{B}\mathbf{u} \quad \forall \mathbf{u}\in{\cal V}
$$

* Representação matricial

$$
[\mathbf{A} + \mathbf{B}] = [\mathbf{A}] + [\mathbf{B}]
$$

## 2)  A multiplicação de um tensor $\mathbf{A}$ por um escalar $\alpha$ é o tensor $\alpha\mathbf{A}$:

* Definição:

$$
(\alpha \mathbf{A})\mathbf{u} = \mathbf{A}(\alpha\mathbf{u})  \quad \forall \alpha\in\mathbb{R}
$$

* Representação matricial

$$
[\alpha\mathbf{A}] = \alpha [\mathbf{A}]
$$

## 3) O produto  $\mathbf{A}\mathbf{B}$ entre os  tensores  $\mathbf{A}$ e $\mathbf{B}$ é o tensor:

* Definição

$$
(\mathbf{A}\mathbf{B})(\mathbf{u}) = \mathbf{A}(\mathbf{B}(\mathbf{u}))
$$

* Representação matricial

$$
[\mathbf{A}\mathbf{B}] = [\mathbf{A}][\mathbf{B}]
$$

# Cuidado:   

Em geral:

$$\boxed{\mathbf{A}\mathbf{B}\neq \mathbf{B} \mathbf{A}}$$

($=$ só em casos especiais!!!!)

#  $\mathbf{A}^\top$: o transposto do tensor $\mathbf{A}$

## 1)  Definição:

$$
\mathbf{A}(\mathbf{u})\cdot\mathbf{v} = \mathbf{u}\cdot \mathbf{A}^\top(\mathbf{v})
$$
para quaisquer vetores $\mathbf{u}$ e $\mathbf{v}$

## 2) Relação entre as matrizes de $\mathbf{A}$ e $\mathbf{A}^\top$:

$$
[\mathbf{A}^\top] = [\mathbf{A}]^\top
$$
ou seja:
$$
[\mathbf{A}^\top] =
\left[
\begin{array}{lcr}
A_{11} & A_{21} & A_{31} \cr
A_{12} & A_{22} & A_{32} \cr
A_{13} & A_{23} & A_{333}
\end{array}
\right]
$$

## 3) Propriedades

$$
\begin{array}{lcr}
(\mathbf{A} + \mathbf{B})^\top = \mathbf{A}^\top + \mathbf{B}^\top \cr
(\mathbf{A}\mathbf{B})^\top = \mathbf{B}^\top\mathbf{A}^\top \cr
((\mathbf{A})^\top)^\top = \mathbf{A}
\end{array}
$$
Em termos matriciais:
$$
\begin{array}{lcr}
[\mathbf{A} + \mathbf{B}]^\top = [\mathbf{A}]^\top + [\mathbf{B}]^\top \cr
[\mathbf{A}\mathbf{B}]^\top = [\mathbf{B}]^\top[\mathbf{A}]^\top \cr
[[\mathbf{A}]^\top]^\top = [\mathbf{A}]
\end{array}
$$

Determinante de um tensor e tensores inversíveis

## Considere um tensor $\mathbf{S}$  qualquer
###  1) Existe tensor $\mathbf{A}$ tal que
$$
\mathbf{v} = \mathbf{S}\mathbf{u} \implies  \mathbf{u} = \mathbf{A}\mathbf{v}
$$
para quaisquer $\mathbf{u}$ e $\mathbf{v}$?

####  R1) Sim: $\mathbf{S}$ é inversível (não singular) e $\mathbf{A}=\mathbf{S}^{-1}$

$$
\mathbf{v} = \mathbf{S}\mathbf{u} \implies  \mathbf{u} = \mathbf{S}^{-1}\mathbf{v}
$$

#### R2)  Não: $\mathbf{S}$ não é inversível (singular)

### 2) Como saber se $\mathbf{S}$ é inversível ou não?

#### R) Através do determinante de $\mathbf{S}$
$$
\left\{
\begin{array}{lcr}
\det\mathbf{S} \neq 0: \quad \mathbf{S} \quad \mbox{inversivel}\cr
\det\mathbf{S} = 0 : \quad \mathbf{S} \quad \mbox{nao inversivel}
\end{array}
\right.
$$

### 3) Como calcular o determinante de um tensor?
$$
\det\mathbf{S} = \det\mathbf{[S]}
$$






# Cálculo do determinante


## 1) Desenvolvimento do determinante de $\mathbf{a}=[\mathbf{A}]$ segundo a $i$-ésima linha

$$
\det\mathbf{a}\, = \sum_{j=1}^n (-1)^{i+j} a_{ij} \det\mathbf{M}_{ij}
$$
onde  $\mathbf{M}_{ij}$ é
a matriz de ordem $(n − 1) \times (n − 1)$ obtida mediante a omissão
da $i$-ésima linha e da $j$-ésima coluna da matriz $\mathbf{a}$

### 1.1) Exemplo: desenvolvimento pela linha 1

$$
\det \left[
\begin{array}{l}
1 & 2 & 3 \cr
4 & 5 & 6 \cr
7 & 8 & 9
\end{array}
\right] = (-1)^{1+1} \times 1\times \det \left[
\begin{array}{l}
 5 & 6 \cr
 8 & 9
\end{array}
\right] + (-1)^{1+2} \times 2 \times \det \left[
\begin{array}{l}
 4 & 6 \cr
 7 & 9
\end{array}
\right] +  (-1)^{1+3} \times 3 \times \det \left[
\begin{array}{l}
 4 & 5 \cr
 7 & 8
\end{array}
\right] \\
= 1 (45-48) - 2 (36-42) + 3 (32-35) = -3 + 12 - 9 = 0
$$

## 2) Método eliminação de Gauss: calcula o determinante da matriz escalonada
$$
\left[
\begin{array}{l}
1 & 2 & 3 \cr
4 & 5 & 6 \cr
7 & 8 & 9
\end{array}
\right] \xrightarrow[L_3- 7L_1]{L_2-4L_1}
\left[
\begin{array}{l}

0 & -3 & -6 \cr
0 & -6 & -12
\end{array}
\right] \xrightarrow{L_3-2L_2} \left[
\begin{array}{l}
1 & 2 & 3 \cr
0 & -3 & -6 \cr
0 & 0  &  0
\end{array}
\right]
$$

$$
\det \left[
\begin{array}{l}
1 & 2 & 3 \cr
4 & 5 & 6 \cr
7 & 8 & 9
\end{array}
\right] = \det \left[
\begin{array}{l}
1 & 2 & 3 \cr
0 & -3 & -6 \cr
0 & 0  &  0
\end{array}
\right] = 1 \times (-3) \times 0 = 0
$$

# Tensores ortogonais

###  1) Um tensor $\mathbf{Q}$ é dito ortogonal quando preserva o produto interno

$$
\mathbf{Q}\mathbf{u}\cdot \mathbf{Q}\mathbf{v} = \mathbf{u}\cdot\mathbf{v}
$$
### para quaisquer $\mathbf{u}$ e $\mathbf{v}$. De forma equivalente:
$$
\begin{array}{lcr}
\mathbf{Q}\mathbf{Q}^\top = \mathbf{Q}^\top \mathbf{Q} = \mathbf{I} \cr
\mathbf{Q}^\top  = \mathbf{Q}^{-1}
\end{array}
$$

### 2) Determinante de um tensor ortogonal

 Se  $\mathbf{Q}$ é ortogonal

$$
\det\mathbf{Q} =  1 \quad\mbox{ou}\quad \det\mathbf{Q} =  -1
$$

De fato:
$$
\begin{array}{lcr}
\mathbf{Q}^\top \mathbf{Q} = \mathbf{I}\implies \underbrace{\det(\mathbf{Q}^\top \mathbf{Q})}_{\det(\mathbf{Q}^\top)\det(\mathbf{Q})} = \underbrace{\det\mathbf{I}}_{1} \implies \cr  \implies (\det\mathbf{Q})^2 = 1
\end{array}
$$

$$

$$

 (u_2 v_3 - u_3 u_2) \mathbf{e_1} +  (u_3 v_1 - u_1 v_3) \mathbf{e_2} +  (u_1 v_2 - u_2 v_1) \mathbf{e_3}

(\mathbf{u}\times\mathbf{v})\cdot\mathbf{w} =  (\mathbf{w}\times\mathbf{u})\cdot\mathbf{v} =  (\mathbf{v}\times\mathbf{w})\cdot\mathbf{u}

#Autovalores, autovetores, e  invariantes principais de um tensor

## 1) Autovalores e autovetores do tensor $\mathbf{S}$:

$$
\mathbf{S}\mathbf{e} = \lambda \mathbf{e}, \quad |\mathbf{e}|=1
$$

$\lambda$:  autovalor  

$\mathbf{e}$: autovetor correspondente

Note que:
$$
\mathbf{S}\mathbf{e} = \lambda \mathbf{e} \implies (\mathbf{S} - \lambda\mathbf{I})\mathbf{e} = \mathbf{0} \implies (\mathbf{S} - \lambda\mathbf{I})\quad\mbox{singular} \implies \det(\mathbf{S} - \lambda\mathbf{I})=0
$$

## 2)  Obtenção dos autovalores de $\mathbf{S}$:  

 Solução do *polinômio característico*

$$
\lambda^3 - i_1(\mathbf{S}) \lambda^2 + i_2(\mathbf{S}) \lambda -  i_3(\mathbf{S})  = 0
$$

## 3) Invariantes principais de $\mathbf{S}$
$$
\begin{array}{lcr}
i_1(\mathbf{S}) = \mbox{tr}\mathbf{S} \cr  i_2(\mathbf{S}) =\frac{1}{2}((\mbox{tr}\mathbf{S})^2 - \mbox{tr}\mathbf{S^2}) \cr i_3(\mathbf{S}) = \det\mathbf{S}
\end{array}
$$

## 4) Teorema de Caley-Hamilton
$$
\mathbf{S}^3 - i_1(\mathbf{S}) \mathbf{S}^2 + i_2(\mathbf{S}) \mathbf{S} -  i_3(\mathbf{S})\mathbf{I}  = \mathbf{0}
$$

## 5) E quando $\mathbf{S}$ é simétrico?

### a) O polinômio característico possui três raizes reais

$$
\mathbf{S}\mathbf{e}_1 = \lambda_1 \mathbf{e}_1, \quad \mathbf{S}\mathbf{e}_2 = \lambda_2 \mathbf{e}_2, \quad \mathbf{S}\mathbf{e}_3 = \lambda_3 \mathbf{e}_3
$$

### b) Os autovetores correspondentes são ortogonais entre si


$$
\mathbf{e}_1\cdot \mathbf{e}_2 = \mathbf{e}_1\cdot \mathbf{e}_3 = \mathbf{e}_2\cdot \mathbf{e}_3 = 0
$$

![Alt text](https://upload.wikimedia.org/wikipedia/commons/c/c7/Mohr_Circle.svg)

```python
"""
Mohr circle in 3D.
@author: Nicolás Guarín-Zapata
@date: May 2020
"""
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigvalsh
from matplotlib import rcParams

rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16


def plot_mohr3d(S):
    r"""Plot 3D Mohr circles."""

    S3, S2, S1 = eigvalsh(S)

    R_maj = 0.5*(S1 - S3)
    cent_maj = 0.5*(S1+S3)

    R_min = 0.5*(S2 - S3)
    cent_min = 0.5*(S2 + S3)

    R_mid = 0.5*(S1 - S2)
    cent_mid = 0.5*(S1 + S2)

    circ1 = plt.Circle((cent_maj,0), R_maj, facecolor='#cce885', lw=3,
                       edgecolor='#5c8037')
    circ2 = plt.Circle((cent_min,0), R_min, facecolor='w', lw=3,
                       edgecolor='#15a1bd')
    circ3 = plt.Circle((cent_mid,0), R_mid, facecolor='w', lw=3,
                       edgecolor='#e4612d')
    plt.axis('image')
    ax = plt.gca()
    ax.add_artist(circ1)
    ax.add_artist(circ2)
    ax.add_artist(circ3)
    ax.set_xlim(S3 - .1*R_maj, S1 + .1*R_maj)
    #ax.set_ylim(-1.1*R_maj, 1.1*R_maj)
    ax.set_ylim(0, 1.1*R_maj)
    #plt.xlabel(r"$\sigma$", size=18)
    #plt.ylabel(r"$\tau$", size=18)
    plt.xlabel(r"$s_n$", size=18)
    plt.ylabel(r"$s_c$", size=18)
    #plt.savefig('Mohr_circle_3D.svg')
    plt.show()


S = np.array([
    [90,0,95],
    [0,96,0],
    [95,0,0])

plt.figure()
plot_mohr3d(S)

S2 = np.array([
    [1,0,0],
    [0,2,0],
    [0,0,4]])

plt.figure()
plot_mohr3d(S2)
plt.show()
```

#Gradiente de um **campo escalar** ($\nabla f$?)

 $f:\cal{E}\rightarrow \mathbb{R}$
$$
 \mathbf{x}\in\cal{E}\mapsto f(\mathbf{x}) \in\mathbb{R}
$$

($\mathbf{x}$: ponto do espaço Euclideano)
#### P1)  $f$ diferenciável em $\mathbf{x}_0\in \cal{E}$ ?

#### R1)   $f$  admite aproximação linear na vizinha de $\mathbf{x}_0$
$$
f(\mathbf{x}_0+\mathbf{h}) \approx f(\mathbf{x}_0) +  \underbrace{Df(\mathbf{x}_0)[\mathbf{h}]}_{(\mbox{termo linear em $\mathbf{h}$})}
$$
para qualquer vetor  $\mathbf{h}\in \cal{V}$  suficiente pequeno, ou seja, $\mathbf{x}_1 = \mathbf{x}_0 + \mathbf{h}$ suficientemente próximo de $\mathbf{x}_0$.



#### De forma mais precisa:

$$
f(\mathbf{x}_0+\mathbf{h}) = f(\mathbf{x}_0) +  Df(\mathbf{x}_0)[\mathbf{h}] + r(\mathbf{h})
$$
tal que
$$
\lim_{|\mathbf{h}|\rightarrow {0}} \frac{r(\mathbf{h})}{|\mathbf{h}|} = 0  \iff r(\mathbf{h}) =  o(\mathbf{h})
$$


 De forma compacta:

*  $f$ é diferenciável em $\mathbf{x}_0$:
$$
\boxed{f(\mathbf{x}_0+\mathbf{h}) = f(\mathbf{x}_0) +  Df(\mathbf{x}_0)[\mathbf{h}] + o(\mathbf{h})}
$$

* Derivada de $f$ em $\mathbf{x}_0$ avaliada no vetor $\mathbf{h}$
$$
Df(\mathbf{x}_0)[\mathbf{h}]
$$

Note que  **$Df(\mathbf{x}_0)$ é uma transformação linear que leva vetores em escalares**
$$
 \mathbf{h}\in{\cal{V}}\mapsto Df(\mathbf{x}_0)[\mathbf{h}] \in\mathbb{R},
$$
 ou seja,  **$Df(\mathbf{x}_0)$ é um funcional linear**. Portanto, existe $\mathbf{a}\in {\cal V}$ tal que:

$$
Df(\mathbf{x}_0)[\mathbf{h}] = \mathbf{a}\cdot\mathbf{h}
$$

##  $\nabla{f}$: gradiente do campo escalar $f$

$$
Df(\mathbf{x})[\mathbf{h}] = \boxed{\nabla{f}(\mathbf{x})\cdot\mathbf{h}  = \lim_{\alpha\to 0} \frac{{f}(\mathbf{x}+\alpha\mathbf{h})-{f}(\mathbf{x})}{\alpha}}
$$

($f$ diferenciável em todos os pontos)



#Gradiente de um **campo vetorial**  ($\nabla\mathbf{f}$ ? )


Campo vetorial: $\mathbf{f}:\cal{E}\rightarrow {\cal V}$
$$
 \mathbf{x}\in\cal{E}\mapsto \mathbf{f}(\mathbf{x}) \in{\cal V}
$$

($\mathbf{x}$: ponto do espaço Euclideano)
#### P1)   $\mathbf{f}$ diferenciável em $\mathbf{x}_0\in \cal{E}$ ?

#### R1)   $\mathbf{f}$  admite aproximação linear na vizinha de $\mathbf{x}_0$
$$
\mathbf{f}(\mathbf{x}_0+\mathbf{h}) \approx \mathbf{f}(\mathbf{x}_0) +  \underbrace{D\mathbf{f}(\mathbf{x}_0)[\mathbf{h}]}_{(\mbox{termo linear em $\mathbf{h}$})}
$$
para qualquer vetor  $\mathbf{h}\in \cal{V}$  suficiente pequeno, ou seja, $\mathbf{x}_1 = \mathbf{x}_0 + \mathbf{h}$ suficientemente próximo de $\mathbf{x}_0$.



#### De forma mais precisa:

$$
\mathbf{f}(\mathbf{x}_0+\mathbf{h}) = \mathbf{f}(\mathbf{x}_0) +  D\mathbf{f}(\mathbf{x}_0)[\mathbf{h}] + \mathbf{r}(\mathbf{h})
$$
tal que
$$
\lim_{|\mathbf{h}|\rightarrow {0}} \frac{|\mathbf{r}(\mathbf{h})|}{|\mathbf{h}|} = 0  \iff \mathbf{r}(\mathbf{h}) =  o(\mathbf{h})
$$


 De forma compacta:

*  $\mathbf{f}$ é diferenciável em $\mathbf{x}_0$:
$$
\boxed{\mathbf{f}(\mathbf{x}_0+\mathbf{h}) = \mathbf{f}(\mathbf{x}_0) +  D\mathbf{f}(\mathbf{x}_0)[\mathbf{h}] + o(\mathbf{h})}
$$

* Derivada de $\mathbf{f}$ em $\mathbf{x}_0$ avaliada no vetor $\mathbf{h}$
$$
D\mathbf{f}(\mathbf{x}_0)[\mathbf{h}]
$$

Note  **$D\mathbf{f}(\mathbf{x}_0)$ é uma transformação linear que leva vetores em vetores**

$$
 \mathbf{h}\in{\cal{V}}\mapsto D\mathbf{f}(\mathbf{x}_0)[\mathbf{h}] \in{\cal V},
$$
 ou seja,  **$D\mathbf{f}(\mathbf{x}_0)$ é um tensor**. Portanto, existe $\mathbf{A}\in {\cal Lin}$ tal que:

$$
D\mathbf{f}(\mathbf{x}_0)[\mathbf{h}] = \mathbf{A}\mathbf{h}
$$

##  $\nabla\mathbf{f}$: gradiente do campo vetorial $\mathbf{f}$

( $\mathbf{f}$ diferenciável em todos os pontos)

$$
D\mathbf{f}(\mathbf{x})[\mathbf{h}] = \nabla\mathbf{f}(\mathbf{x})\mathbf{h}=\boxed{\nabla\mathbf{f}(\mathbf{x})\mathbf{h}  = \lim_{\alpha\to 0} \frac{\mathbf{f}(\mathbf{x}+\alpha\mathbf{h})-\mathbf{f}(\mathbf{x})}{\alpha}}
$$




# Operadores diferenciais sobre campos: gradient, divergente $\mbox{div}$, rotacional $\mbox{rot}$ ($\mbox{curl}$ ), e laplaciano $\triangle$


## 1) O  gradiente do campo escalar $\varphi: {\cal D}\subset {\cal E} \rightarrow \mathbb{R}$ é o campo de vetores  

$$\nabla\varphi: {\cal D}\subset {\cal E} \rightarrow {\cal V}$$ tal que
$$
\boxed{D\varphi(\mathbf{x})[\mathbf{u}] = \nabla\varphi(\mathbf{x})\cdot \mathbf{u}}
$$
para todo vetor  $\mathbf{u}$

## 2) O gradiente do  campo de vetores $\mathbf{v}: {\cal D}\subset {\cal E} \rightarrow {\cal V}$ é o campo de tensores
$$
\nabla\mathbf{v}: {\cal D}\subset {\cal E} \rightarrow {\cal Lin}$$ tal que
$$
\boxed{D\mathbf{v}(\mathbf{x})[\mathbf{u}] = \nabla\mathbf{v}(\mathbf{x})\mathbf{u}}
$$
para todo vetor  $\mathbf{u}$

## 3) O divergente do campo de vetores $\mathbf{v}: {\cal D}\subset {\cal E} \rightarrow {\cal V}$ é o campo de  escalares
$$
\mbox{div} \mathbf{v}: {\cal D}\subset {\cal E} \rightarrow \mathbb{R}$$ tal que
$$
\boxed{\mbox{div} \mathbf{v}= \mbox{tr}(\nabla\mathbf{v})}
$$

## 4) O rotacional do campo de vetores $\mathbf{v}: {\cal D}\subset {\cal E} \rightarrow {\cal V}$ é o campo de vetores
$$
\mbox{rot} \mathbf{v}: {\cal D}\subset {\cal E} \rightarrow {\cal V}$$ tal que
$$
\boxed{ (\nabla\mathbf{v} - (\nabla\mathbf{v})^\top)  \mathbf{a}= \mbox{rot}\mathbf{v}\times\mathbf{a}}
$$
para qualquer vetor $\mathbf{a}$

 \mathbf{F}=\nabla \mathbf{f} = \mathbf{I} + \nabla\mathbf{u}\iff \nabla\mathbf{u}= \mathbf{F} - \mathbf{I}
 $$

- Decomposição polar de $\mathbf{F}$
$$
\mathbf{F}=\mathbf{R}\mathbf{U}
$$
  * Estiramento ($\mathbf{U}$) e rotação ($\mathbf{R}$)

* Pequenas deformações  $|\nabla\mathbf{u}|\ll 1$
Neste caso:
  - $\mathbf{U} \approx  \mathbf{I} + \mathbf{E}$ e $\mathbf{R} \approx  \mathbf{W}$ onde o tensor simétrico
$$\boxed{\mathbf{E}=\frac{\nabla\mathbf{u} + \nabla\mathbf{u}^\top}{2}}$$ é o *tensor de deformação infinitesimal* e o tensor antissimétrico   $$\mathbf{W}=\frac{\nabla\mathbf{u} - \left(\nabla\mathbf{u}\right)^\top}{2}$$
é o *tensor de rotação infinitesimal*.  


Note que:
$$
\boxed{\nabla\mathbf{u} =\mathbf{E} + \mathbf{W}}
 $$



Considere  $\mathbf{x}$ and $\bar{\mathbf{x}}$ as localizações do mesmo ponto material nas configurações indeformada e  deformada do corpo material ${\cal B}$
  * Deformação  ($\mathbf{f}$) e deslocamento ($\mathbf{u}$)
  $$
  \mathbf{x}\mapsto \bar{\mathbf{x}}=\mathbf{f}(\mathbf{x})=\mathbf{x}+\mathbf{u}(\mathbf{x})
  $$

Em um sistema de coordenadas cartesiano:
  $$
  \left\{
  \begin{array}{lcr}
  \mathbf{x}=(x,y,z), \quad \bar{\mathbf{x}}=(\bar{x},\bar{y},\bar{z}) \cr \mathbf{f}(\mathbf{x})=(f_x(x,y,z),f_y(x,y,z),f_z(x,y,z)) \cr \mathbf{u}(\mathbf{x})=(u(x,y,z),v(x,y,z),w(x,y,z))
  \end{array}
  \right. $$

Um cisalhamento simples é uma deformação que pode ser expressa da seguinte forma:
 $$
  \left\{
  \begin{array}{lcr}
  \bar{x} = f_x(x,y,x) = x + \gamma y \cr
  \bar{y}  = f_y(x,y,z) = y \cr
  \bar{z} = f_x(x,y,z) = z
  \end{array}
  \right. \implies
  \left\{
  \begin{array}{lcr}
  u(x,y,x) =  f_x(x,y,z) - x = \gamma y \cr
  v(x,y,z)  = f_y(x,y,z) - y  = 0 \cr
  w(x,y,z) = f_z(x,y,z) -z = 0
  \end{array}
  \right.
$$
- $\gamma$: quantidade de cisalhamento

Neste caso a matriz de $\nabla\mathbf{u}$ é dada por
$$
[\nabla\mathbf{u}]=
\left[
\begin{array}{lcr}
\frac{\partial u}{\partial x} & \frac{\partial u}{\partial y} & \frac{\partial u}{\partial z} \cr
\frac{\partial v}{\partial x} & \frac{\partial v}{\partial y} & \frac{\partial v}{\partial z} \cr
\frac{\partial w}{\partial x} & \frac{\partial w}{\partial y} & \frac{\partial w}{\partial z}
\end{array}  
\right]= \left[
\begin{array}{lcr}
0 & \gamma & 0 \cr
0 & 0 & 0\cr
0 & 0 & 0
\end{array}  
\right]
$$
Portanto
$$[\mathbf{E}]=\left[
\begin{array}{lcr}
0 & \frac{\gamma}{2} & 0 \cr
\frac{\gamma}{2} & 0 & 0\cr
0 & 0 & 0
\end{array}  
\right]\quad\mbox{e} \quad
[\mathbf{W}]=\left[
\begin{array}{lcr}
0 & \frac{\gamma}{2} & 0 \cr
-\frac{\gamma}{2} & 0 & 0\cr
0 & 0 & 0
\end{array}  
\right]
$$


Para pequenas deformações é necessário que $|\nabla\mathbf{u}|\ll 1$, o que no caso do cisalhemente equivale a
$$
|\nabla\mathbf{u}|=\sqrt{0^2+0^2+0^2+\gamma^2 + \dots + 0^2}=|\gamma| \ll 1
$$

##1) Corpo material ${\cal B}$

1. Conjunto formado por pontos materiais
$$
\mathbf{p} \in {\cal B}:\quad \mbox{ponto material}
$$
2. Ocupa regiões do espaço Euclideano 3D
3. Identificado com a região ocupada em uma **configuração de referência**

(pontos materiais idetificados com pontos do espaço)

##2) Deformação  de ${\cal B}$ (mudança de lugar e de forma):   mapeamento

$$
\mathbf{f}:{\cal B} \rightarrow {\cal E}
$$


1. $\mathbf{x}=\mathbf{f}(\mathbf{p})$: lugar ocupado pelo ponto material $\mathbf{p}$ após a deformação $\mathbf{f}$

2. $\mathbf{f}({\cal B})$: região ocupada por ${\cal B}$ na **configuração deformada**

### Requisitos sobre $\mathbf{f}$:

1. Um lugar no espaço só pode ser ocupado por um único ponto material

2. $\mathbf{f}({\cal B})$ é um-a-um entre pontos materiais e lugares no espaço

3. $\nabla\mathbf{f}= \nabla_{\mathbf{p}}\mathbf{f}$ admite inversa
$$
\det\nabla\mathbf{f}\neq 0
$$
### Sem perda de generalidade, vamos exigir  que
$$
\boxed{\det\nabla\mathbf{f} > 0}
$$

## 3) Deslocamento de um ponto material

$$
\boxed{\mathbf{u}(\mathbf{p}) = \mathbf{f}(\mathbf{p}) - \mathbf{p}}.
$$
  - Pequenas deformações
  $$
  \boxed{|\nabla\mathbf{u}|\ll 1}
  $$

# O gradiente de deformação $\mathbf{F}=\nabla\mathbf{f}$ (leitura adicional)

$$
\mathbf{f}(\mathbf{p}) = \mathbf{f}(\mathbf{q}) + \mathbf{F}(\mathbf{q})(\mathbf{p}-\mathbf{q}) + o(\mathbf{p}-\mathbf{q})
$$
## 1) Deformação homogênea:  $\mathbf{F}$ é  constante

$$
\mathbf{f}(\mathbf{p}) = \mathbf{f}(\mathbf{q}) + \mathbf{F}(\mathbf{p}-\mathbf{q})
$$
onde $\mathbf{p}$ e $\mathbf{q}$ são pontos materiais quaisquer

1.  Deformação rígida: deformação homogênea onde $\mathbf{F}=\mathbf{R}$ (rotação)

## 2)  Na vizinha de um ponto $\mathbf{q}$ e dentro de um erro $o(\mathbf{p}-\mathbf{q})$ qualquer deformação se comporta como homogênea
$$
\boxed{\mathbf{f}(\mathbf{p}) \approx \mathbf{f}(\mathbf{q}) + \mathbf{F}(\mathbf{p}-\mathbf{q})}
$$

## 3) Decomposição polar de $\mathbf{F}$ (em cada ponto)

$$
\mathbf{F}=\mathbf{RU}=\mathbf{VR}
$$

$\mathbf{R}$: tensor rotação; $\mathbf{U}=\sqrt{\mathbf{C}}$: tensor estiramento direto; $\mathbf{V}=\sqrt{\mathbf{B}}$: tensor estiramento esquerdo

$\mathbf{C}=\mathbf{F}^\top\mathbf{F}$: tensor de Cauchy-Green direito

$\mathbf{B}=\mathbf{F}\mathbf{F}^\top$: tensor de Cauchy-Green esquerdo

## 4) Autovalores e autovetores de $\mathbf{U}$: estiramentos principais e direções principais de estiramento

# Interpretação de $\mathbf{F}$  no caso de deformação homogênea: estiramento e   cisalhamento (leitura adicional)


## 1) Razão das distâncias entre os pontos materiais  $\mathbf{q}$ e $\mathbf{p}$,  após e antes da deformação:

$$
\begin{array}{lcr}
\frac{|\mathbf{f}(\mathbf{q})- \mathbf{f}(\mathbf{p})|}{|\mathbf{q}-\mathbf{p}|} &=& \cr
&=& \frac{|\mathbf{F}(\mathbf{q} - \mathbf{p})|}{|\mathbf{q}-\mathbf{p}|} \cr
&=& \sqrt{\frac{\mathbf{F}(\mathbf{q} - \mathbf{p})\cdot \mathbf{F}(\mathbf{q} - \mathbf{p})}{|\mathbf{q}-\mathbf{p}|^2}} \cr
&=& \sqrt{\mathbf{F}\mathbf{e}\cdot \mathbf{F}\mathbf{e}}= |\mathbf{F}\mathbf{e}|
\end{array}
$$
onde
$
\mathbf{e} = \frac{\mathbf{q}-\mathbf{p}}{|\mathbf{q}-\mathbf{p}|}
$ (vetor unitário)

## 2) Estiramento de uma linha material paralela ao vetor unitário $\mathbf{e}$

$$
\begin{array}{lcr}
\lambda(\mathbf{e}) &=& |\mathbf{F}\mathbf{e}| \cr
&=&|\mathbf{RU}\mathbf{e}|\cr
&=&|\mathbf{U}\mathbf{e}|
\end{array}
$$

1. Estiramentos máximo e mínimo e direções correspondentes

$$
\max_{|\mathbf{e}|=1}\lambda(\mathbf{e}) = \mbox{maior}(\lambda_1, \lambda_2, \lambda_3) \quad \min_{|\mathbf{e}|=1}\lambda(\mathbf{e}) = \mbox{menor}(\lambda_1, \lambda_2, \lambda_3)
$$
Supondo $\lambda_1\ge \lambda_2\ge \lambda_3$:
$$
\max_{|\mathbf{e}|=1}\lambda(\mathbf{e}) = \lambda_1 \quad \min_{|\mathbf{e}|=1}\lambda(\mathbf{e}) = \lambda_3
$$

## 3) Angulo formado  por  linhas materiais deformadas  

$$
\begin{array}{lcr}
\cos\theta = \frac{(\mathbf{f}(\mathbf{q})- \mathbf{f}(\mathbf{p}))\cdot (\mathbf{f}(\mathbf{r})- \mathbf{f}(\mathbf{p}))}{|\mathbf{f}(\mathbf{q})-\mathbf{f}(\mathbf{p})||\mathbf{f}(\mathbf{r})-\mathbf{f}(\mathbf{p})|} &=& \cr
&=& \frac{\mathbf{F}\mathbf{e}\cdot \mathbf{F}\mathbf{g}}{\lambda(\mathbf{e})\lambda(\mathbf{g})} \cr
&=& \frac{\mathbf{U}\mathbf{e}\cdot \mathbf{U}\mathbf{g}}{\lambda(\mathbf{e})\lambda(\mathbf{g})}
\end{array}
$$
$
\mathbf{e} = \frac{\mathbf{q}-\mathbf{p}}{|\mathbf{q}-\mathbf{p}|}
$ e $
\mathbf{g} = \frac{\mathbf{r}-\mathbf{p}}{|\mathbf{r}-\mathbf{p}|}
$ (vetores unitários)

1. Ângulo de cisalhamento entre as linhas materiais deformadas  inicialmente perpendiculares entre si

$$
\gamma(\mathbf{e},\mathbf{g}) = \frac{\pi}{2} - \theta(\mathbf{e},\mathbf{g})
$$

$$
\mathbf{f}(\mathbf{p}) = \mathbf{p} + \mathbf{u}(\mathbf{p})

$$

$$

$$
\mathbf{C} = \mathbf{I} + \underbrace{\nabla\mathbf{u} + (\nabla\mathbf{u})^\top}_{2\mathbf{E}} + (\nabla\mathbf{u})^\top \nabla\mathbf{u}
$$

$$
\mathbf{B} = \mathbf{I} + \underbrace{\nabla\mathbf{u} + (\nabla\mathbf{u})^\top}_{2\mathbf{E}} + \nabla\mathbf{u} (\nabla\mathbf{u})^\top

$$
\begin{array}{lcr}
\mathbf{C} &\approx&  \mathbf{I} + 2 \mathbf{E} \cr
\mathbf{B} &\approx& \mathbf{I} + 2 \mathbf{E} \cr
\mathbf{U} &\approx&  \mathbf{I} + \mathbf{E} \cr
\mathbf{\mathbf{V} } &\approx& \mathbf{I} +\mathbf{E} \cr

* $\mathbf{W}$: tensor de rotações  infinitesimal

Considere um corpo material ${\cal B}$
  - sistema mecânico interesse
  - sólido deformável tridimensional
-  $\mathbf{x}\in {\cal B}$: ponto material ou partícula
  - ${\cal P}$: parte de ${\cal B}$

Diagrama de corpo livre para ${\cal B}$
-  sistema de forças e momentos atuando sobre ${\cal B}$
  - forças externas ao sólido  ${\cal B}$

Diagrama de corpo livre para ${\cal P}$  
  - sistemas de forças e momentos atuando sobre ${\cal P}$
    - forças externas à parte  ${\cal P}$
    - forças externas e internas à ${\cal B}$

# Balanço de forças e o tensor de tensões

A resultante do sistema de forças que atua sobre uma parte arbitrária  ${\cal P}$ de ${\cal B}$ deve ser ser nula:
$$
\boxed{
\left\{
\begin{array}{lcr}
\boldsymbol{F}_R({\cal P})= 0 \cr
\mbox{para toda parte}\quad{\cal P}\quad\mbox{de}\quad {\cal B}
\end{array}
\right.
}
$$
Hipóteses:

1) O sistema de forças que atua sobre  ${\cal P}$ é formado por  dois tipos forças,  *forças de contato* e *forças de ação à distância*.

Logo:
$$
\boxed{\boldsymbol{F}_R({\cal P})= \boldsymbol{F}_R^C({\cal P})+\boldsymbol{F}_R^D({\cal P})}
$$

onde $\boldsymbol{F}_R^C({\cal P})$ e $\boldsymbol{F}_R^D({\cal P})$ representam as resultantes das forças de contato e das forças de ação à distância que atuam sobre a parte ${\cal P}$.

2) As resultantes das forças de contato e das forças de ação à distância que atuam sobre ${\cal P}$ admitem a seguinte representação:

$$
\boxed{\boldsymbol{F}_R^C({\cal P}) = \int_{\partial_i {\cal P}} \mathbf{T}(\mathbf{x})\mathbf{n}(\mathbf{x})\; dA + \int_{\partial_e {\cal P}} \mathbf{s}(\mathbf{x})\; dA} \quad\&\quad \boxed{\boldsymbol{F}_R^D({\cal P}) = \int_{\cal P} \mathbf{b}(\mathbf{x})\;dV}
$$
onde:
- ${\partial \cal P}$ representa a fronteira ou contorno de ${\cal P}$
  - $\partial_i{\cal P}$: parte da fronteira de ${\cal P}$ no interior de ${\cal B}$
  - $\partial_e{\cal P}$: parte da fronteira de ${\cal P}$ que pertence a fronteira de ${\cal B}$
  - $\partial{\cal P}=\partial_i{\cal P}$: parte totalmente no interior de ${\cal B}$
- $\mathbf{T}$ é o campo de tensões
  - $\mathbf{T}(\mathbf{x})$ tensor de tensões no ponto material $\mathbf{x}$
- $\mathbf{n}$ é o campo de vetores unitários normais à superfície ${\partial \cal P}$ apontando para fora da região ${\cal P}$
  - $\mathbf{n}(\mathbf{x})$ é o vetor unitário normal à superfície $\partial {\cal P}$ em $\mathbf{x}$
- $\mathbf{s}(\mathbf{x})$: vetor força  por unidade de área atuando na fronteira de ${\cal B}$
- $\mathbf{b}$ é  densidade volumétrica  das forças de corpo  
  - $\mathbf{b}(\mathbf{x})$ é a força de corpo por unidade de volume em $\mathbf{x}$

Observações sobre $\mathbf{T}\mathbf{n}$:
-   densidade superficial das forças de contato
-  vetor tensão que atua no plano definido  pelo vetor unítario  $\mathbf{n}$




# Balanço de forças na forma local

Com as hipóteses acima podemos reescrever o balanço de forças da seguinte forma:
$$
\left\{
\begin{array}{lcr}
\int_{\partial {\cal P}} \mathbf{T}(\mathbf{x})\mathbf{n}(\mathbf{x})\; dA + \int_{\cal P} \mathbf{b}(\mathbf{x})\;dV = \mathbf{0} \cr
\mbox{para toda parte}\quad{\cal P}\quad\mbox{no interior de}\quad {\cal B}.
\end{array}
\right.
$$
Usando o teorema da divergência:
$$
\left\{
\begin{array}{lcr}
\int_{ {\cal P}} (\mbox{div}\,\mathbf{T}(\mathbf{x})+ \mathbf{b}(\mathbf{x}))\;dV = \mathbf{0} \cr
\mbox{para toda parte}\quad{\cal P}\quad\mbox{de}\quad {\cal B}.
\end{array}
\right.
$$
A declaração acima é equivalente  ao balanço de forças na *forma forte*
$$
\boxed{
\left\{
\begin{array}{lcr}
\mbox{div}\,\mathbf{T}(\mathbf{x})+ \mathbf{b}(\mathbf{x})= \mathbf{0} \cr
\mbox{para todo}\quad\mathbf{x}\quad\mbox{no interior de}\quad {\cal B}.
\end{array}
\right.
}
$$

Já para pontos na fronteira de ${\cal B}$ pode-se concluir que

$$
\boxed{
\left\{
\begin{array}{lcr}
\mathbf{T}(\mathbf{x})\mathbf{n}(\mathbf{x})=\mathbf{s}(\mathbf{x}) \cr
\mbox{para todo}\quad\mathbf{x}\;\;\;\mbox{na fronteira de}\;\;\; {\cal B}
\end{array}
\right.}
$$

\right.
}
$$

$$
\left[
  \begin{array}{lcr}

  \end{array}
\right] = \lambda (\varepsilon_{11}+\varepsilon_{22}+\varepsilon_{33})
\left[
\begin{array}{lcr}
  1 & 0 & 0 \cr
  0 & 1 & 0 \cr
  0 & 0 & 1
  \end{array}
 \right] + 2\mu \left[
  \begin{array}{lcr}
  \varepsilon_{11} & \varepsilon_{12} & \varepsilon_{13} \cr

  \end{array}
\right]
$$

- Passo 2:  Relacionar as componentes de tensão com as componentes de deformação:
$$
\left\{
\begin{array}{lcr}

\sigma_{22} = \lambda  (\varepsilon_{11}+\varepsilon_{22}+\varepsilon_{33}) + 2 \mu \varepsilon_{22} \cr
\sigma_{33} = \lambda  (\varepsilon_{11}+\varepsilon_{22}+\varepsilon_{33}) + 2 \mu \varepsilon_{33} \cr
\sigma_{12} = 2\mu \varepsilon_{12} \cr

\sigma_{23} = 2\mu \varepsilon_{23}
\end{array}  
\right.  
$$

- Passo 3: Resolver o sistema acima para as componentes de deformação
  - Observando que
$$
 \sigma_{11}+\sigma_{22}+\sigma_{33}= (3\lambda+2\mu)(\varepsilon_{11}+\varepsilon_{22}+\varepsilon_{33})
$$
e agrupando termos:

$$
\left.
\begin{array}{lcr}
\varepsilon_{11} = \frac{\lambda+\mu}{\mu(3\lambda+2\mu)}\left(\sigma_{11} - \frac{1}{2(\lambda+\mu)} (\sigma_{22}+\sigma_{33})\right)\cr
\varepsilon_{22} = \frac{\lambda+\mu}{\mu(3\lambda+2\mu)}\left(\sigma_{22} - \frac{1}{2(\lambda+\mu)} (\sigma_{11}+\sigma_{33})\right)\cr

\varepsilon_{13} = \frac{1}{2 \mu}\sigma_{13} \cr
\varepsilon_{23} = \frac{1}{2 \mu}\sigma_{23}

\left\{
\begin{array}{lcr}
\varepsilon_{11} = \frac{1}{E}\left(\sigma_{11} - \nu (\sigma_{22}+\sigma_{33})\right)\cr
\varepsilon_{22} = \frac{1}{E}\left(\sigma_{22} - \nu (\sigma_{11}+\sigma_{33})\right)\cr
\varepsilon_{33} = \frac{1}{E}\left(\sigma_{33} - \nu (\sigma_{11}+\sigma_{22})\right)\cr
\gamma_{12} = 2 \varepsilon_{12} = \frac{1}{G}\sigma_{12} \iff \sigma_{12}= G \gamma_{12} \cr
\gamma_{13}=2\varepsilon_{13} = \frac{1}{ G}\sigma_{13}  \iff \sigma_{13}= G \gamma_{13} \cr
\gamma_{23}=2\varepsilon_{23} = \frac{1}{ G}\sigma_{23}  \iff \sigma_{23}= G \gamma_{23}
\end{array}  
\right.

# Decomposição aditiva dos tensores  $\mathbf{E}$ e $\mathbf{T}$

- Observação: qualquer tensor $\mathbf{A}$ admite a representação (verifique)
$$
\boxed{\mathbf{A} =\mathbf{A}_M + \mathbf{A}_D}, \quad{\mbox{com}} \quad \boxed{\mathbf{A}_M =  \frac{\mbox{tr}\mathbf{A}}{3} \mathbf{I}}\quad\mbox{e}\quad \boxed{\mathbf{A}_D = \left(\mathbf{A}-  \frac{\mbox{tr}\mathbf{A}}{3} \mathbf{I} \right)}
$$
onde:
  - $\mathbf{A}_M$ é a *parte média* ou volumétrica de $\mathbf{A}$
  - $\mathbf{A}_D$ é a *parte desviadora* $\mathbf{A}$
    - Note que
    $$
    \boxed{\mbox{tr}(\mathbf{A}_M)=\mbox{tr}(\mathbf{A})}\quad \boxed{\mbox{tr}(\mathbf{A}_D)=0} \quad \boxed{\mathbf{A}_M\cdot\mathbf{A}_D = 0}
    $$
- Para quaisquer tensores $\mathbf{A}$ e $\mathbf{B}$:
$$
\boxed{\mathbf{A}_M\cdot\mathbf{B}_D = \mathbf{A}_D\cdot\mathbf{B}_M= 0}
$$


### 1)  Decomposição de  $\mathbf{E}$

$$
\mathbf{E} = \mathbf{E}_M + \mathbf{E}_D
$$

- $\mathbf{E}_M = (\varepsilon_V/3)\mathbf{I}$: deformação volumétrica  sem mudança de forma (dilatação)
- $\mathbf{E}_D$: deformação com mudança de forma a volume constante (distorção)

### 2)  Decomposição de  $\mathbf{T}$

$$
\mathbf{T} = \mathbf{T}_M + \mathbf{T}_D
$$

- $\mathbf{T}_M = \sigma_V \mathbf{I}$: tensão hidroestática
- $\mathbf{T}_D$: tensão desviadora

### 3) Relação constitutiva

$$
\mathbf{T}=\lambda(\mbox{tr}\mathbf{E})\mathbf{I} + 2\mu \mathbf{E} \iff \boxed{\mathbf{T}_M = \kappa \mathbf{E}_M}  \quad \boxed{\mathbf{T}_D = 2 \mu \mathbf{E}_D}
$$

# A  energia de deformação elástica

A energia armazenada por um  sólido elástico linear ${\cal B}$ após deformação é dada por:
$$
E({\cal B}) = \int_{\cal B} \psi(\mathbf{E}(\mathbf{x}))\,dV
$$

- $ \psi(\mathbf{E}(\mathbf{x}))$ é a energia de deformação elástica por unidade de volume (densidade de energia)  em $\mathbf{x}$


### 1) A densidade de energia $\psi$ para um sólido elástico linear e isotrópico

$$
\boxed{\psi(\mathbf{E}) = \frac{\lambda}{2} \left(\mbox{tr}\mathbf{E}\right) ^2 + \mu |\mathbf{E}|^2}
$$
Pode ser mostrado que:
$$
\mathbf{T}=\frac{\partial \psi(\mathbf{E})}{\partial \mathbf{E}} = \lambda (\mbox{tr}\mathbf{E})\mathbf{I} + 2\mu \mathbf{E}
$$


Observações:

- Considerando que
$$
\mathbf{T}\cdot\mathbf{E} = \lambda (\mbox{tr}\mathbf{E})\mathbf{I} + 2\mu \mathbf{E})\cdot\mathbf{E} = \lambda (\mbox{tr}\mathbf{E})\mathbf{I}\cdot\mathbf{E} + 2\mu \mathbf{E}\cdot\mathbf{E}  = \lambda (\mbox{tr}\mathbf{E})^2 + 2\mu |\mathbf{E}|^2
$$
podemos escrever
$$
\boxed{\psi = \frac{1}{2}\mathbf{T}\cdot\mathbf{E}}
$$
-  Analogamente para uma mola elástica  linear:
$$
E(x) = \frac{1}{2} Kx^2  \iff F = \frac{d E}{d x} = Kx  \iff E = \frac{Fx}{2}
$$


### 2) Energias de mudança de volume e distorção:

Observando que
$$
\mathbf{T}\cdot\mathbf{E} = \mathbf{T}_M\cdot \mathbf{E}_M + \mathbf{T}_D\cdot \mathbf{E}_D = \sigma_V \varepsilon_V + \mathbf{T}_D\cdot \mathbf{E}_D
$$
e lembrando que
$$
\sigma_V = \kappa \varepsilon_v \quad \mathbf{T}_D=2\mu \mathbf{E}_D,
$$
podemos concluir que:
$$
\boxed{\psi(\mathbf{E}) = \frac{1}{2} \kappa \varepsilon_V^2 + \mu |\mathbf{E}_D|^2}
$$
- Densidade de energia associada à mudança de volume
$$
\boxed{\psi_V(\varepsilon_V) = \frac{1}{2} \kappa \varepsilon_V^2}
$$
- Densidade de energia associada à mudança de forma (energia de distorção)
$$
\boxed{\psi_D(\mathbf{E}_D) =  \mu |\mathbf{E}_D|^2}
$$

Note que podemos escrever $\psi_V$ e $\psi_D$ como:

$$
\boxed{\psi_V(\mathbf{T}_M) = \frac{1}{2 \kappa}  \sigma_M^2} \quad
\boxed{\psi_D(\mathbf{T}_D) =  \frac{1}{\mu} |\mathbf{T}_D|^2}
$$

# Região elástica e critérios de resistência

O comportamento de um sólido ${\cal  B}$  é elástico desde que:

$$
\boxed{f(\mathbf{T}(\mathbf{x})) < f_L}
$$

para todo ponto material $\mathbf{x}$ de ${\cal B}$,  onde $f_L$ é o valor limite de $f$.

De agora em diante vamos omitir o ponto $\mathbf{x}$

### 1) Região elástica no espaço das tensões é formada pelos tensores $\mathbf{T}$ tais  que
$$
f(\mathbf{T}) < f_L
$$

-  Para materiais isotrópicos:
$$
f(\mathbf{T})= f(\sigma_1,\sigma_2,\sigma_3)
$$

- Tensão equivalente $\sigma_{eq}$
$$
f(\sigma_1,\sigma_2,\sigma_3) = f(\sigma_{eq},0,0) \iff \sigma_{eq} = g(\sigma_1,\sigma_2,\sigma_3)
$$

- Região elástica em termos da tensão equivente:
$$
\boxed{\sigma_{eq} < \sigma_L}
$$
onde
$\sigma_L$ é a tensão limite  em ensaio de tração  

### 2) Critérios de Rankine, Tresca, e von Mises

Definir um critério de resitência é definir a função $f$:

- Rankine (máxima tensão normal)

$$
\boxed{f(\mathbf{T}) = \sigma_1 \iff  \sigma_{eq}=\sigma_1}
$$

- Tresca (máxima tensão de cisalhamento)

$$
\boxed{f(\mathbf{T}) = \frac{\sigma_1-\sigma_3}{2} \iff  \sigma_{eq}=\sigma_1-\sigma_3}
$$

- von Mises (máxima energia de distorção)

$$
\boxed{f(\mathbf{T}) = \frac{1}{\mu} |\mathbf{T}_D|^2  \iff  \sigma_{eq}=\sqrt{\frac{3}{2}}|\mathbf{T}_D|^2}
$$

# Calculando  as constantes $A$ e $B$ a partir das condições de contorno $\sigma_r(a)=-p_i$ e $\sigma_r(b)=-p_e$ (via Python)

```python
#@title
from IPython.display import Math, Latex
from sympy import *
A, B = symbols('A, B')
a, b = symbols('a, b', positive=True)
pi, pe = symbols(r'p_i, p_e')
r = symbols('r')
sigmar, sigmateta=symbols('sigma_r,sigma_theta')
#
# Tensões radial e circunferencial
#
sigmar = Function('sigma_r')(r)
sigmat = Function('sigma_theta')(r)
#
sigmar= A - B/(r**2)
display(Math(r'\sigma_r(r) = {}'.format(latex(sigmar))))
sigmat= A + B/(r**2)
display(Math(r'\sigma_\theta(r) = {}'.format(latex(sigmat))))
```

A ideia agora é obter $A$ e $B$ considerando as condições de contorno para o problema térmico:
$$
\sigma_r(a)=-p_i \quad \sigma_r(b)=-p_e
$$

As equações a serem resolvidas para $A$ e $B$ são implementadas da sequinte forma:

```python
#@title
lhs1 = sigmar.subs([(r,a)])
rhs1 = -pi
eq1 = Eq(lhs1,rhs1)
#display(eq1)
#
lhs2 = sigmar.subs([(r,b)])
rhs2 = -pe
eq2 = Eq(lhs2,rhs2)
#display(eq2)
```

Agora vamos encontrar $A$ e $B$:

```python
#@title
soln, = linsolve([eq1,eq2],[A,B])

Asoln = soln[0]
Bsoln = soln[1]
AsolS=simplify(Asoln)
BsolS=simplify(Bsoln)

display(Math(r'A = {}'.format(latex(AsolS))))
display(Math(r'B = {}'.format(latex(BsolS))))
```

Agora vamos plotar $\sigma_r$, $\sigma_\theta$, e $\sigma_{eq}^{Tresca}$  (admitindo $\sigma_z \ge 0$), normalizados pela pressão interna $p_i=p$,   em função da coordenada radial $r$, normalizada pelo raio interno $a> 0$,  adotando:  $p_e=0$  e $b=2a$.

```python
#@title
#import nump as np
sigmarr=sigmar.subs([(A,AsolS),(B,BsolS)])
sigmatt=sigmat.subs([(A,AsolS),(B,BsolS)])
#
vna=1; vnb=2*vna; vnpi=1; vnpe=0
#
#
psigmarr=sigmarr.subs([(pi,vnpi),(pe,vnpe),(a,vna), (b,vnb)])
psigmatt=sigmatt.subs([(pi,vnpi),(pe,vnpe),(a,vna), (b,vnb)])
l_sigmar=lambdify(r,psigmarr)
l_sigmat=lambdify(r,psigmatt)
```

 B= (\bar p_i-\bar p_e)\frac{a^2b^2}{b^2-a^2}
 \end{array}
 }
 \right\} \quad\mbox{para}\quad a > 0 \quad\mbox{e}\quad  \left.
 \boxed{
 \begin{array}{lcr}
 A= - \bar p_e  \cr
 B= 0
 \end{array}
 }
 \right\} \quad\mbox{para}\quad a = 0
 $$
 onde:
 $$
 \left\{
   \boxed{
 \begin{array}{lcr}
 \bar p_i = - \frac{\rho\omega^2 a^2}{8}(3+\nu) \cr
 \bar p_e = - \frac{\rho\omega^2 b^2}{8}(3+\nu)
 \end{array}
   }
 \right.
 $$

 - Estado de deformações
 $$
 [\mathbf{E}(\mathbf{x})]=
 \left[
   \begin{array}{lcr}
   \varepsilon_r(r) & 0 & 0 \cr
   0 & \varepsilon_{\theta}(r) & 0 \cr
   0 & 0 & \varepsilon_{z}(r)
   \end{array}
 \right]  
 $$
onde:
$$
\left\{
\boxed{
\begin{array}{lcr}
\varepsilon_{r} = \frac{1}{E}\left(\sigma_{r} - \nu (\sigma_{\theta}+\sigma_{z})\right)\cr
\varepsilon_{\theta} = \frac{1}{E}\left(\sigma_{\theta} - \nu (\sigma_{r}+\sigma_{z})\right)\cr
\varepsilon_{z} = -\frac{\nu}{E} (\sigma_{r}+\sigma_{\theta})
\end{array}
}
\right.
$$  
  - Note que $\varepsilon_z$ **não**  é constante!

- Campo de deslocamentos
  - $u_\theta=0$
  - $u_r$ e $u_z$:
$$
\left\{
\begin{array}{lcr}
\frac{\partial u_r}{\partial r}= \varepsilon_r  \cr
 \frac{u_r}{r}= \varepsilon_\theta   \cr
 \frac{\partial u_z}{\partial z} =\varepsilon_z
\end{array}
\right. \implies
\left\{
  \boxed{
\begin{array}{lcr}
 \frac{u_r(r)}{r}= \varepsilon_\theta(r)  
\end{array}
  }
\right.
$$

# Derivação dos resultados

### Método semi-inverso:

#### **Passo 1**.  Hipóteses sobre a solução:
  - Simetria de revolução:
      - Deslocamento radial $u_r$ depende apenas de $r$
      - Deslocamento circunferencial $u_\theta$ é nulo
  - Estado plano de tensão
  $$
  \sigma_z=\tau_{rz}=\tau_{\theta z}=0,
  $$


#### **Passo 2**.   Consequências das hipóteses adotadas
  - Usando a relação deformação-deslocamento:
  $$
  \varepsilon_r = \frac{d u_r}{d r}, \quad \varepsilon_\theta = \frac{u_r}{r}, \quad \varepsilon_{r\theta} = 0
  $$
  - Usando as relações constituvas:
  $$
 [\mathbf{T}]=
 \left[
   \begin{array}{lcr}
   \sigma_r & 0 & 0 \cr
   0 & \sigma_{\theta} & 0 \cr
   0 & 0 & 0
   \end{array}
 \right]  
 $$
onde:
$$
\left\{
\begin{array}{lcr}
\sigma_r = \lambda \left(\frac{d u_r}{d r} + \frac{u_r}{r} + \varepsilon_z \right) + 2 \mu \frac{d u_r}{d r} \cr
\sigma_\theta = \lambda \left(\frac{d u_r}{d r} + \frac{u_r}{r} + \varepsilon_z \right) + 2 \mu \frac{u_r}{ r} \cr
\sigma_z = \lambda \left(\frac{d u_r}{d r} + \frac{u_r}{r} + \varepsilon_z \right) + 2 \mu \varepsilon_z
\end{array}
\right.
$$
Logo:
$$
 \varepsilon_z =\frac{-\lambda}{\lambda + 2\mu}\left(\frac{d u_r}{d r} + \frac{u_r}{r}\right)
$$

    - **Pelas equações $\sigma_r$, $\sigma_\theta$, e $\sigma_z$ só dependem de $r$!**

  - Das equações de equilíbrio na presença de forças de corpo:
  $$
  \frac{d \sigma_r}{d r}  + \frac{\sigma_r-\sigma_\theta}{r} + \rho\omega^2 r= 0 \implies   \frac{d}{d r}\left(\frac{1}{r}\frac{d}{d r}\left({u_r}{r}\right)\right)=-\frac{1-\nu^2}{E}\rho\omega^2 r  \implies \\ \boxed{u_r = C_1 r + \frac{C_2}{r} - \frac{1-\nu^2}{E}\rho\omega^2\frac{r^3}{8}}
  $$
onde $C_1$ e $C_2$ são constantes de integração a serem determinadas!

\left\{

\begin{array}{lcr}
\sigma_r = A - \frac{B}{r^2} - \frac{\rho\omega^2 r^2}{8}(3+\nu) \cr
\sigma_\theta = A + \frac{B}{r^2} - \frac{\rho\omega^2 r^2}{8}(1+2\nu)
\end{array}}
\right.
$$
onde $A$ e $B$ são constantes a serem determinadas.

# Exercício

Considere um cilindro de parede grossa fechado de raio interno $a$ e raio externo $b$ submetido a um pressao interna  $p$.

1. Obtenha as distribuições das tensões radial, circunfercial, e equivalente de Tresca

2. Obtenha a tensão equivalente de  Tresca máxima

3.  Use o resultado para encontrar a relação entre a $a$, $b$, e $p$ e a tensão admissível $\sigma_{adm}$. Em particular, obtenha:
  - i) a pressão máxima admíssivel suportada por um cilindro fechado de raio interno  de $a$ e  raio externo $b$;  
  - ii) a espessura necessária para o que um cilindro fechado de raio interno $a$ suporte uma pressão interna   $p$.


**Solução:**

1. Sabemos que o estado de tensões é definido por:
$$
\sigma_r(r) = A - \frac{B}{r^2}, \quad \sigma_\theta(r) = A + \frac{B}{r^2}, \quad \sigma_z  = A,
$$
onde $A$ e $B$ são obtidas através do uso  das condições de contorno $\sigma_r(a)=-p$ e $\sigma_r(b)=0$. Logo:
$$
A = p \frac{a^2}{b^2-a^2} \quad\mbox{e}\quad B=p \frac{a^2b^2}{b^2-a^2}.
$$
Avaliando as expressões acima podemos concluir que
$$
\sigma_1(r) = \sigma_\theta(r), \quad \sigma_2(r) = \sigma_z, \quad\mbox{e}\quad \sigma_3(r) = \sigma_r(r).
$$
Logo, a tensão equivalente de Tresca  avaliada em  $a \le r\le b$ é dada por:
$$
\sigma_{eq}^T(r) = \sigma_1(r)-\sigma_3(r) = 2\frac{B}{r^2} =2p \frac{a^2b^2}{b^2-a^2}\frac{1}{r^2}
$$
2. A tensão equivalente de Tresa máxima é dada por:
$$
(\sigma_{eq}^T)_{max}:=\max_{a \le r\le b} \sigma_{eq}^T(r) = \max_{a \le r\le b} \left(2\frac{B}{r^2}\right) =2p \frac{b^2}{b^2-a^2}  
$$
3. A tensão admissível $\sigma_{adm}$ é definida por
$$
\sigma_{adm}=\frac{\sigma_L}{n},
$$
onde $\sigma_L$ é a tensão limite do material e $n> 1$ o coeficiente de segurança. A análise desse item é baseada na relação
$$
(\sigma_{eq}^T)_{max} = \sigma_{adm} \iff 2p \frac{b^2}{b^2-a^2}  = \sigma_{adm}
$$
A expressão acima fornece a relação entre  as dimensões ($a$ e $b$) e o carregamento ($p$) para que o cilindro trabalhe de modo seguro.
  - i) Pressão admissível máxima:
  $$
  p_{adm} =  \frac{b^2-a^2}{2 b^2} \sigma_{adm}
  $$
  - ii) Espessura:
  $$
  h = b-a, \quad\mbox{onde}\quad  b= a \sqrt{ \frac{\sigma_{adm}}{\sigma_{adm}-2p}}.
  $$
  Note que a equação acima só tem solução se
  $$
  \boxed{p < \frac{\sigma_{adm}}{2}}
  $$


  **Como projetar um cilindro para suportar  uma pressão interna
  $
  p > \frac{\sigma_{adm}}{2}
  $ ?**

# Ajuste com interferência

**Definição do problema**:

Considere dois corpos cilindrícos:
1. Componente 1: raio externo $R_1$
2. Componente 2: raio interno $R_2$
$$
R_1 > R_2 \iff \delta = R_1 - R_2 > 0
$$

Como montar o  componente 1 no componente 2?
-  Ajuste com interferência
- Dimensões do conjunto após a montagem:
$$
a < c < b
$$

**Perguntas:**
1. Pressão de contato $p_c$ após a montagem
2. Estado de tensões após a montagem
3. Estado de tensões em serviço

**Respostas:**

Após a montagem, o componente 1 estará submetido a um pressão interna $p_c$. Já  o componente 2 estará submetido a um pressão externa $p_c$. Como estimar $p_c$?
- Passo 1:
$$
\left.
\begin{array}{lcr}
R_1 \rightarrow R_1 + u_1(R_1) = c \cr
R_2 \rightarrow R_2 + u_2(R_2) = c
\end{array}
\right\} \implies  u_2(R_2)- u_1(R_1) = \delta
$$
- Passo 2:
$$
\left.
\begin{array}{lcr}
 u_1(R_1) \approx  c \varepsilon_\theta^1(c) \cr
u_2(R_2) \approx  c \varepsilon_\theta^2(c)
\end{array}
\right\} \implies  c (\varepsilon_\theta^2(c) - \varepsilon_\theta^1(c) )  = \delta
$$
- Passo 3. Usando a relações constitutiva
$$
\varepsilon_\theta = \frac{1}{E} (\sigma_\theta - \nu (\sigma_r + \sigma_z))
$$
para cada componente após a montagem:
$$
\left.
\begin{array}{lcr}
 \varepsilon_\theta^1(c) = \frac{1}{E_1} (\sigma_\theta^1(c) - \nu_1 \sigma_r^1(c) ) \cr
 \varepsilon_\theta^2(c) = \frac{1}{E_2} (\sigma_\theta^2(c) - \nu_2\sigma_r^2(c) )
\end{array}
\right\}
$$
- Passo 4. Escrever as tensões em função de $p_c$

   - Componente 1: $a \le r \le c$
   $$
\sigma_r^1(r) = A_1 - \frac{B_1}{r^2} \quad \& \quad \sigma_\theta^1(r) = A_1 + \frac{B_1}{r^2}
   $$
   onde $A_1$ e $B_1$ são obtidos através da resolução do sistema
  $$
\left\{
\begin{array}{lcr}
 \sigma_r^1(a)=0 \cr
\sigma_r^1(c)=-p_c
\end{array}
\right. \implies
\left\{
\boxed{
\begin{array}{l}
A_1 = - \frac{p_c c^2}{c^2- a^2} \cr
B_1 = - \frac{p_c a^2 c^2}{c^2- a^2}
\end{array}
}
\right.
$$

  - Componente 2: $c \le r \le b$
   $$
\sigma_r^2(r) = A_2 - \frac{B_2}{r^2} \quad \& \quad \sigma_\theta^2(r) = A_2 + \frac{B_2}{r^2}
   $$
   onde $A_2$ e $B_2$ são obtidos através da resolução do sistema
  $$
\left\{
\begin{array}{lcr}
 \sigma_r^2(c)=-p_c \cr
\sigma_r^2(b)=0
\end{array}
\right.  \implies
\left\{
\boxed{
\begin{array}{l}
A_2 =  \frac{p_c c^2}{b^2- c^2} \cr
B_2 =  \frac{p_c b^2 c^2}{b^2- c^2}
\end{array}
}
\right.
$$

- Passo 5. Obter $p_c$ através da resulação da equação:
$$
 \boxed{\frac{1}{E_2} (\sigma_\theta^2(c) - \nu_2 \sigma_r^2(c))  - \frac{1}{E_1} (\sigma_\theta^1(c) - \nu_1 \sigma_r^1(c))   = \frac{\delta}{c}}
$$

* Mostre que para  cilindros de mesmo material ($E_1=E_2=E$ e $\nu_1=\nu_2=\nu)$:
$$
\boxed{
\delta = \frac{2 p_c c^3}{E} \frac{(b^2-a^2)}{(b^2-c^2)(c^2-a^2)}}
$$


O estado de tensões no conjunto após a montagem é dado por
$$
\boxed{
\sigma_r^m(r) =
\left\{
\begin{array}{lcr}
 \sigma_r^1(r),  \quad a \le r \le c, \cr
\sigma_r^2(r), \quad c \le r \le b,
\end{array}
\right.} \quad
\boxed{
\sigma_\theta^m(r) =
\left\{
\begin{array}{lcr}
 \sigma_\theta^1(r),  \quad a \le r \le c, \cr
\sigma_\theta^2(r), \quad c \le r \le b,
\end{array}
\right. }
\quad\&\quad \boxed{\sigma_z^m = 0}.
$$

As tensões  no componente composto  quando  em  serviço  é são dadas por:
$$
\sigma_r = \sigma_r^m + \sigma_r^s, \quad \sigma_\theta = \sigma_\theta^m + \sigma_\theta^s, \quad\mbox{e}\quad \sigma_z = \sigma_z^s
$$
onde as $\sigma_r^s$, $\sigma_\theta^s$, e $\sigma_z^s$  são calculados como se estivéssemos trabalhando com um cilindro de raio interno $a$ e raio externo $b$. Isso nos casos dos componentes serem do mesmo material.

- Escolha ótima para $c$:

    - Pontos críticos:   superfícies internas dos cilindros interno e  externo

    $$
    
    - Escolha ótima de $c$ minimiza $ \sigma_{eq}^{max}$

\right.
}  
$$

# Tensões térmicas

## Mecanismos de deformação: elástico e térmico

- Decomposição aditiva dos tensões de deformações em suas parcelas elástica $\mathbf{E}_e$ e térmica $\mathbf{E}_T$:
$$
\mathbf{E} = \mathbf{E}_e + \mathbf{E}_T
$$

-  Equação constitutiva para a parte térmica:
$$
\mathbf{E}_T = \alpha (\theta-\theta_R) \mathbf{I}
$$
  - $\alpha$: coeficiente de expansão térmica
  - $\theta$: temperatura
  - $\theta_R$: temperatura de referência

- Equação constitutiva para o tensor de tensões $\mathbf{T}$:

$$
\mathbf{T}=\mathbb{C}[\mathbf{E}_e] = \lambda (\mbox{tr}\mathbf{E}_e)\mathbf{I} + 2\mu \mathbf{E}_e = \lambda (\mbox{tr}\mathbf{E})\mathbf{I} + 2\mu \mathbf{E} \color{blue}{-(3\lambda + 2\mu)\alpha\triangle\theta\mathbf{I}}
$$  
onde
$$
\triangle \theta = \theta - \theta_R
$$
é a variação de temperatura.

### **Problema:**

Um  tubo cilíndrico longo com raio interno $a$ e raio externo $b$ transporta  um  fluido quente com temperatura $T_a$. A superfície externa do tubo possui temperatura $T_b$. Considerando que as superfícies interna e externa do tubo são livres de tração,  assumindo estado plano de deformação deformação de deformação plana, e supondo que a distribuição de temperatura no tubo seja dada por
$$
T(r)=\frac{T_a\log\Big(\frac{r}{b}\Big)−T_b\log\Big(\frac{r}{a}\Big)}{\log\Big(\frac{a}{b}\Big)} = T_b + \frac{T_a-T_b}{\log\left(\frac{b}{a}\right)}\log\left(\frac{b}{r}\right)
$$

1) Obtenha as componentes de tensão $\sigma_r$, $\sigma_\theta$, e $\sigma_z$ na tubulação;

2) Encontrar a expressão para a tensão equivalente de Von Mises na tubuluação. Onde o valor máximo é atingido?

3) Obtenha a diferenta de temperatura $T_a-T_b$ que provocará falha (escoamento) em tubulação de aço.


### **Solução:**

Aspectos comuns  das equações que governo para qualquer problema com simetria de revolução do tipo considerado:
- Relação deformação-deslocamento
$$
\boxed{
\begin{array}{l}
\varepsilon_r = \frac{d u}{d r} \cr
\varepsilon_\theta = \frac{ u}{r}
\end{array}
}
$$
onde $u$ é o deslocamento radial, função de $r$ apenas.

- Equação de equilíbrio
$$
\boxed{
\frac{d \sigma_r}{d r} + \frac{\sigma_r-\sigma_\theta}{r} + b_r = 0
}
$$

- Equações constitutivas
$$
\boxed{
\begin{array}{lcr}
\varepsilon_r = \frac{1}{E} \left(\sigma_r - \nu (\sigma_\theta + \sigma_z)\right) + {\color{blue}{\alpha\triangle T}} \cr
\varepsilon_\theta = \frac{1}{E} \left(\sigma_\theta - \nu (\sigma_r + \sigma_z)\right) + {\color{blue}{\alpha\triangle T}} \cr
\varepsilon_z = \frac{1}{E} \left(\sigma_z - \nu (\sigma_r + \sigma_\theta)\right) + {\color{blue}{\alpha\triangle T}}
\end{array}
} \iff
\boxed{
\begin{array}{lcr}
\sigma_r = \lambda (\varepsilon_r + \varepsilon_\theta + \varepsilon_z) + 2\mu \varepsilon_r  - \color{blue}{(3\lambda + 2\mu) \alpha\triangle T} \cr
\sigma_\theta = \lambda (\varepsilon_r + \varepsilon_\theta + \varepsilon_z) + 2\mu \varepsilon_\theta  - \color{blue}{(3\lambda + 2\mu) \alpha\triangle T} \cr
\sigma_z = \lambda (\varepsilon_r + \varepsilon_\theta + \varepsilon_z) + 2\mu \varepsilon_z  - \color{blue}{(3\lambda + 2\mu) \alpha\triangle T}
\end{array}
}
$$
onde $\alpha$ é o coeficiente de dilatação térmica e $\triangle T$ a variação de temperatura com respeito a uma temperatura de referência.

- Condições de contorno
$$
\boxed{
\begin{array}{lcr}
\sigma_r(a)=-p_i \cr
\sigma_r(b)=-p_e
\end{array}
}
$$

Aspectos específicos do caso considerado:
- Estado plano de deformação
$$
\boxed{
\varepsilon_z = 0
}
$$
- Superfícies internas livres de tensão
$$
\boxed{
p_i=p_e=0
}
$$
- Ausência de forças de corpo
$$
\boxed{b_r = 0}
$$

- Temperatura do de referência é a temperatura do ambiente
$$
\triangle T = T - T_b = \frac{T_a-T_b}{\log\left(\frac{b}{a}\right)}\log\left(\frac{b}{r}\right)
$$

Agora estamos pontos para iniciar a solução do nosso problemas. Vamos proceder de acordo com os seguintes passos:

- Passo 1: Usar as relações deformação-deslocamento e  constitutivas para escrever
$$
\boxed{
\begin{array}{lcr}
\sigma_r = \lambda \left( \frac{du}{dr} + \frac{u}{r}\right) + 2\mu \frac{du}{dr} - (3\lambda+2\mu)\alpha\triangle T \cr
\sigma_\theta = \lambda \left( \frac{du}{dr} + \frac{u}{r}\right) + 2\mu \frac{u}{dr} - (3\lambda+2\mu)\alpha\triangle T
\end{array}
}
$$

- Passo 2: Substituir  as equações acima na equação de equilíbrio para obter
$$
(\lambda+2\mu) \left(\frac{d^2 u}{d r^2} +  \frac{d}{d r}\left(\frac{u}{r}\right)\right) = (3\lambda+2\mu)\alpha \frac{d \triangle T}{d r}  \iff \frac{d^2 u}{d r^2} +  \frac{d}{d r}\left(\frac{u}{r}\right)  = \frac{d}{d r}\left(\frac{1}{r}\frac{d}{d r}\left({u}{r}\right)\right) = \frac{3\lambda+ 2\mu}{\lambda + 2 \mu}\alpha \frac{d \triangle T}{d r}
$$
Ou seja:
$$
\boxed{
 \frac{d}{d r}\left(\frac{1}{r}\frac{d}{d r}\left({u}{r}\right)\right)  = \alpha\beta \frac{d \triangle T}{d r}}, \quad \mbox{onde}\quad  \boxed{
\beta = \frac{3\lambda+ 2\mu}{\lambda + 2 \mu}
}
$$
Mostre que:
$$
\boxed{
\beta = \frac{3\lambda+ 2\mu}{\lambda + 2 \mu} =  \frac{{1+\nu}}{1-\nu}
}
$$

- Passo 3: Integrar a equação acima e obter

$$
 \frac{d}{d r}\left(\frac{1}{r}\frac{d}{d r}\left({u}{r}\right)\right)  = \alpha\beta \frac{d \triangle T}{d r} \implies\frac{d}{d r}\left({u}{r}\right) = C_1 r  + \alpha\beta \triangle T r  \\
 \implies \boxed{u = \frac{C_1}{2} r + \frac{C_2}{r} + \frac{ \alpha\beta}{r}\int_a^r \triangle T(r^*)r^* \, dr^*}
$$
ou
$$
u = u_h + u_p \quad\mbox{onde}\quad
\left\{
\begin{array}{lcr}
u_h = \frac{C_1}{2} r + \frac{C_2}{r} \cr
u_p =  \frac{\alpha\beta}{r} \int_a^r \triangle T(r^*)r^* \, dr^*
\end{array}
\right.
$$

Note que:
$$
\left\{
\begin{array}{lcr}
\frac{d u_h}{d r}=  \frac{C_1}{2} - \frac{C_2}{r^2} \cr
\frac{u_h}{r}=  \frac{C_1}{2} + \frac{C_2}{r^2}
\end{array}
\right. \implies \frac{d u_h}{d r} + \frac{u_h}{r} = C_1
$$

$$
\left\{
\begin{array}{lcr}
\frac{d u_p}{d r}= \alpha\beta \left(T - \frac{1}{r^2} \int_a^r \triangle T(r^*) r^*\; dr^*\right)\cr
\frac{u_p}{r}= \alpha\beta \left(\frac{1}{r^2} \int_a^r \triangle T(r^*) r^*\; dr^*\right)
\end{array}
\right. \implies \frac{d u_p}{d r} + \frac{u_p}{r} = \alpha\beta \triangle T
$$
- Passo 4: Escrever $\sigma_r$ e $\sigma_\theta$ usando a expressão para $u$ obtida acima
$$
\left.
\begin{array}{lcr}
\sigma_r = \lambda\left( C_1  + \frac{d u_p}{d r} +  \frac{u_p}{r}\right) + 2 \mu\left(\frac{C_1}{2} - \frac{C_2}{r^2} + \frac{d u_p}{d r}\right) -(3\lambda+2\mu)\alpha\triangle T \cr
\sigma_\theta = \lambda\left( C_1  + \frac{d u_p}{d r} +  \frac{u_p}{r}\right)  + 2 \mu\left(\frac{C_1}{2} + \frac{C_2}{r^2} + \frac{ u_p}{r}\right) -(3\lambda+2\mu)\alpha\triangle T
\end{array}
\right\}
\implies
\left\{
\begin{array}{lcr}
\sigma_r =  A - \frac{B}{r^2} + \lambda\left( \frac{d u_p}{d r}  + \frac{u_p}{r}\right) + 2 \mu\left( \frac{d u_p}{d r}\right) -(3\lambda+2\mu)\alpha\triangle T \cr
\sigma_\theta = A + \frac{B}{r^2} + \lambda\left( \frac{d u_p}{d r} + \frac{u_p}{r}\right) + 2 \mu\left( \frac{ u_p}{r}\right) -(3\lambda+2\mu)\alpha\triangle T
\end{array}
\right. \implies \\
\left\{
\begin{array}{lcr}
\sigma_r =  A - \frac{B}{r^2} - \frac{2 \mu\alpha\beta}{r^2} \int_a^r \triangle T(r^*) r^*\; dr^*  \cr
\sigma_\theta = A + \frac{B}{r^2} +\frac{2 \mu\alpha\beta}{r^2} \int_a^r \triangle T(r^*) r^*\; dr^* - 2\mu\alpha\beta \triangle T  
\end{array}
\right. \iff
  \boxed{
\left\{
\begin{array}{lcr}
\sigma_r =  A - \frac{B}{r^2} - \frac{ E\alpha}{(1-\nu)r^2} \int_a^r \triangle T(r^*) r^*\; dr^*  \cr
\sigma_\theta = A + \frac{B}{r^2} + \frac{ E\alpha}{(1-\nu)r^2} \int_a^r \triangle T(r^*) r^*\; dr^* -  \frac{ E\alpha}{(1-\nu)} \triangle T  
\end{array}
\right.}
$$
onde
$$
A = (\lambda +\mu) C_1 \quad\mbox{e}\quad  B = 2\mu C_2
$$

  - Note que:
  
- Passo 5: Avaliar as constantes $A$ e $B$
$$
\sigma_r(a) = A - \frac{B}{a^2} - \frac{ E\alpha}{(1-\nu)a^2} \int_a^a \triangle T(r^*) r^*\; dr^*  = A - \frac{B}{a^2}=  0
$$

$$
\sigma_r(b) = A - \frac{B}{b^2} - \frac{ E\alpha}{(1-\nu)b^2} \int_a^b \triangle T(r^*) r^*\; dr^*  =  0
$$
Ou seja:

$$
\left\{
\begin{array}{lcr}
A - \frac{B}{a^2} =  0 \cr
A - \frac{B}{b^2} =  \frac{ E\alpha}{(1-\nu)b^2} \int_a^b T(r^*) r^*\; dr^*
\end{array}
\right.
\iff
\boxed{
\left\{
\begin{array}{lcr}
A =   \frac{ E\alpha}{(1-\nu)(b^2-a^2)} \int_a^b \triangle T(r^*) r^*\; dr^* \cr
B = \frac{a^2 E\alpha}{(1-\nu)(b^2-a^2)} \int_a^b \triangle T(r^*) r^*\; dr^*
\end{array}
\right.
}
$$

- Passo 6: Escrever as tensões em função de $\triangle$ usando as expressões de $A$ e $B$ obtidas no item anterior
$$
\begin{array}{lcr}
\sigma_r = \frac{ E\alpha}{(1-\nu)r^2}\left[ \frac{r^2-a^2}{b^2-a^2}\int_a^b \triangle T(r^*) r^*\; dr^* - \int_a^r \triangle T(r^*) r^*\; dr^*\right]  \cr
\sigma_\theta = \frac{ E\alpha}{(1-\nu)r^2}\left[ \frac{r^2+a^2}{b^2-a^2}\int_a^b \triangle T(r^*) r^*\; dr^* + \int_a^r \triangle T(r^*) r^*\; dr^* - \triangle T r^2\right] \cr
\sigma_z = \frac{2 \nu E\alpha}{(1-\nu)(b^2-a^2)}\int_a^b \triangle T(r^*) r^*\; dr^*  -  \frac{E\alpha}{1-\nu}\triangle T
\end{array}
$$
Agora basta usar a expressão de $\triangle T$  e usar integração por partes  para obter
$$
\int_a^r \triangle T(r^*) r^*\; dr^*  = \frac{T_a-T_b}{\log\left(\frac{b}{a}\right)}\left( \frac{r^2}{2}\log\left(\frac{b}{r}\right) - \frac{a^2}{2}\log\left(\frac{b}{a}\right) -  \frac{r^2-a^2}{4} \right)
$$
e
$$
\int_a^b \triangle T(r^*) r^*\; dr^*  = -\frac{T_a-T_b}{\log\left(\frac{b}{a}\right)}\left( \frac{a^2}{2}\log\left(\frac{b}{a}\right) + \frac{b^2-a^2}{4}\right).
$$
De posse  destes resultados:
$$
\boxed{
\begin{array}{lcr}
\sigma_r = \frac{E\alpha(T_a-T_b)}{2(1-\nu)\log\left(\frac{b}{a}\right)}\left[ - \log\left(\frac{b}{r}\right) - \frac{a^2(r^2-b^2)}{r^2(b^2-a^2)}\log\left(\frac{b}{a}\right) \right]  \cr
\sigma_\theta = \frac{E\alpha(T_a-T_b)}{2(1-\nu)\log\left(\frac{b}{a}\right)}\left[ 1 - \log\left(\frac{b}{r}\right) - \frac{a^2(r^2+b^2)}{r^2(b^2-a^2)}\log\left(\frac{b}{a}\right) \right]  \cr
\sigma_z =\frac{E\alpha(T_a-T_b)}{(1-\nu)\log\left(\frac{b}{a}\right)}\left[ \nu \left(\frac{1}{2}  - \frac{ a^2}{(b^2-a^2)}\log\left(\frac{b}{a}\right)\right) - (1+\nu)\log\left(\frac{b}{r}\right) \right]  
\end{array}
}
$$

To find the shear centre, we consider the idealized geometry:

<img src="https://github.com/jeevanjyoti4/advmechsolids/blob/c79da1ffff45f01bd6c1802b21077c13709962fe/channel_section_exploded.svg?raw=1" width="150">

We set the origin of the $yz$ coordinate system at $O$.

We first determine the second moment of area of the whole idealized section.

```python
import sympy as sym
from IPython.display import Math, Latex

b, h, t = sym.symbols('b, h, t', positive=True)

Iz = sym.Rational(1,12)*t*h**3 + 2*b*t*(sym.Rational(1,2)*h)**2

display(Math(r'I_z = {}'.format(sym.latex(Iz))))
```

Note that in the above we have neglected the term $\displaystyle 2 \left(\frac{1}{12}bt^3 \right)$.

Next, we find the $Q_z$. For this we number the various parts of the idealized section: the top flange (i.e. the top horizontal part) is numbered 1; the web (i.e. the vertical part) is numbered 2; and the bottom flange (i.e. the bottom horizontal part) is numbered 3.

For each of the parts, we use the variable $s_i$ to denote the coordinate along the cross-section.

For $Q_z^{(1)}$, we use the local variable $s_1$ such that $s_1 \in [0,\; b]$ which corresponds to $z \in [-b, \; 0]$ (we are moving from the right of the top flange to the left). Thus, we have $\displaystyle Q_z^{(1)} = \int_0^{s_1} t y \; {\rm d}s_1$, where $y = h/2$ is a constant throughout the top flange.

```python
s1, s2, s3 = sym.symbols('s_1, s_2, s_3')
Qz1 = sym.integrate(sym.Rational(1,2)*h*t,(s1,0,s1))
display(Math(r'Q_z^{{(1)}} = {}'.format(sym.latex(Qz1))))
```

For $Q_z^{(2)}$, we use the local variable $s_2$ such that $s_2 \in [0,\; h]$ which corresponds to $y \in [h/2, \; -h/2]$ (we are moving from top of the web to the bottom). However, we must also account for the contribution from the entire top flange. Thus, we have $\displaystyle Q_z^{(2)} = \left. Q_z^{(1)}\right\vert_{s_1 = b} + \int_0^{s_2} t y \; {\rm d}s_2$, where $y = h/2 - s_2$ varies along the web.

```python
Qz2 = ( Qz1.subs(s1,b) + sym.integrate( (sym.Rational(1,2)*h-s2)*t,(s2,0,s2) ) ).simplify()
display(Math(r'Q_z^{{(2)}} = {}'.format(sym.latex(Qz2))))
```

For $Q_z^{(3)}$, we use the local variable $s_3$ such that $s_3 \in [0,\; b]$ which corresponds to $z \in [0, \; -b]$ (we are moving from the left of the bottom flange to the right). However, now, we must also account for the contribution from the entire top flange and the vertical web. Thus, we have $\displaystyle Q_z^{(3)} = \left. Q_z^{(2)}\right\vert_{s_2 = h} + \int_0^{s_3} t y \; {\rm d}s_3$, where $y = -h/2$ is again a constant for the entire bottom flange. Note that $\left. Q_z^{(2)}\right\vert_{s_2 = h}$ contains within it the contribution from both the top flange and the web.

```python
Qz3 = ( Qz2.subs(s2,h) + sym.integrate(-sym.Rational(1,2)*h*t,(s3,0,s3)) ).simplify()
display(Math(r'Q_z^{{(3)}} = {}'.format(sym.latex(Qz3))))
```

Next, we determine the shear stress distribution along the flanges and web using the general formula: $\displaystyle \sigma_{sx} = \frac{V Q}{I_z t}$, where $V$ denotes the shear force at the cross-section under consideration. For a cantilever beam under loaded at the tip by a point force $P$, we must have $V = P$ constant throughout the length of the beam. Thus we have $\displaystyle \sigma_{sx}^{(i)} = \frac{P Q_z^{(i)}}{I_z t}$.

```python
P = sym.symbols('P')

sigma_sx1 = (P*Qz1/(Iz*t)).simplify()
sigma_sx2 = (P*Qz2/(Iz*t)).simplify()
sigma_sx3 = (P*Qz3/(Iz*t)).simplify()

display(Math(r'\sigma_{{sx}}^{{(1)}} = {}'.format(sym.latex(sigma_sx1))))
display(Math(r'\sigma_{{sx}}^{{(2)}} = {}'.format(sym.latex(sigma_sx2))))
display(Math(r'\sigma_{{sx}}^{{(3)}} = {}'.format(sym.latex(sigma_sx3))))
```

For the shear forces we just need to carry out the integrations over the appropriate sections: $\displaystyle V^{(i)} = \int_0^{s_i} \sigma_{sx}^{(i)} t \; {\rm d}s_i$.

```python
V1 = (sym.integrate(sigma_sx1*t,(s1,0,b))).simplify()
display(Math(r'V^{{(1)}} = {}'.format(sym.latex(V1))))


V2 = sym.integrate(sigma_sx2*t,(s2,0,h)).simplify()
display(Math(r'V^{{(2)}} = {}'.format(sym.latex(V2))))


V3 = (sym.integrate(sigma_sx3*t,(s3,0,b))).simplify()
display(Math(r'V^{{(3)}} = {}'.format(sym.latex(V3))))
```

Before we proceed to finally obtaining the location of the shear centre, it is extremely important to check a couple of things.

__Check 1:__ The magnitude of $V^{(2)}$ must be equal to $P$. This is true, as can be seen above.

__Check 2:__ The magnitude of $V^{(1)}$ and $V^{(3)}$ must be equal because only then the resultant of the horizontal forces in the cross-section will be zero. This is true again, as can be seen above.

Finally, to obtain the location of the shear centre, we determine the torsional moment of $V^{(1)}$ and $V^{(3)}$ about $O$. The sense of both these torsional moments are equal and so their contributions add up.

```python
torsional_moment = V1*h/2 + V3*h/2
display(torsional_moment)
```

This torsional moment must be balanced by the torsional moment due to $P$ which is applied at the shear centre. If the distance of the shear centre from $O$ is taken to be $e$ (along the $z$-axis) then we have:

```python
e = torsional_moment/P
display(Math(r'e = {}'.format(sym.latex(e))))
```

# Viga coluna

Considere uma barra reta de comprimento $L$ submetida a um carregamento no plano $x-y$.

As equações que governam o comportamento de uma viga coluna são as seguintes:
- Equações de equilíbrio
$$
\left\{
\begin{array}{lcr}
N' + f_x = 0 \cr
V' + f_y = 0 \cr
m' + V  \color{blue}{ - Nv'} = 0
\end{array}
\right.
$$
onde $N$ é o esforço normal, $V\equiv V_y$ o esforço cortante, $m\equiv m_z$ o momento fletor, $f_x$ e $f_y$ as cargas distribuidas nas direções $x$ e $y$, e $v$ o deflexão (deslocamento) na direção $y$
  - O sistema de coordenadas é definido na configuração não pertubada (eixo reto) e todas quantidades definidas em relação a este sistema
  - As equações de equilíbrio obtidas usando a configuração pertubada (eixo  curvo)

- Equação constitutiva
$$
m = E I v''
$$  
onde $E$ é o módulo de elasticidade e $I\equiv I_z$ o momento de inércia.
Note que
$$
\theta = v'  \quad (\mbox{Euler-Bernoulli})
$$

Às equações acima devem ser acrescidas as
- Condições de contorno
 - $x=0$
$$
\begin{array}{lcr}
N(0) = -F_x(0) \quad\mbox{ou}\quad u(0)=u_0 \cr
V(0) = -F_y(0) \quad\mbox{ou}\quad v(0)=v_0 \cr
m(0) = -M(0) \quad\mbox{ou}\quad \theta(0)=\theta_0
\end{array}
$$

 - $x=L$
$$
\begin{array}{lcr}
N(L) = F_x(L) \quad\mbox{ou}\quad u(L)=u_L \cr
V(L) = F_y(L) \quad\mbox{ou}\quad v(L)=v_L \cr
m(L) = -M(L) \quad\mbox{ou}\quad \theta(L)=\theta_L
\end{array}
$$
      - Observação: condições de contorno envolvendo o deslocamento na direção $x$ ($u$) não serão usadas no presente desenvolvimento

- Condições de continuidade ($v$ e $\theta$) e de salto ($N$, $V$ e $m$)  qunado necessário

### Equação diferencial para $v$ ($N=-P$ e $f_y=0$)
$$
\boxed{v^{iv} + \kappa^2 v'' = 0},  \quad  \kappa^2 =  \frac{P}{EI}
$$  
A  solução mais geral da equação tem a seguinte forma:

$$
\boxed{
v(x) = C_1 + C_2 x + C_3 \sin \kappa x + C_4 \cos \kappa x
}
$$   
- Problemas envolvendo vigas colunas passam pela avaliação das  constantes acimas através das quatros condições de contorno, duas em $x=0$ e duas em $x=L$,  envolvendo $(v,V)$ e $(m,\theta)$. Para isto, vamos usar as seguintes relações:
$$
\left\{
\begin{array}{lcr}
\theta = v' \cr
m = EI v '' \cr
V = -m' - Pv' = -(EIv''' + Pv')
\end{array}
\right.
$$
Logo:
$$
\boxed{\mathbf{A}(\kappa,L)\mathbf{c}=\mathbf{b}}
$$
onde $\mathbf{c}=(C_1,C_2,C_3,C_4)$ representa as incógnitas do problema, send $\mathbf{A}$ e $\mathbf{b}$ determinados pelos condições de contorno;

## Solução

Vamos considerar o sistema de coordenadas global $(X,Y,X)$ com origem em $A$. Também vamos considerar o sistema local $(x,y,z)$ tal que:

- Barra $AB$: $x=X$, $y=Y$, e $z=Z$;

- Barra $BC$: $x=Z$, $y=Y$, e $z=-X$ (conformidade com a regra da mão direita)


O carregamento aplicado na extremidade livre $C$  é
$$
\mathbf{F}_C = - P \mathbf{e}_Y
$$

i) Obtenção dos esforço cortante momentos fletor e torsor

- Para uma seção $x$ na barra $AB$:
$$
\mathbf{F}_C - \mathbf{F}^i(x) = \mathbf{0} \implies \mathbf{F}^i(x) =  - P \mathbf{e}_Y \implies V_y = - P \mathbf{e}_Y\cdot\mathbf{e}_y = - P \mathbf{e}_Y\cdot\mathbf{e}_Y = - P
$$

$$
\left((L-x)\mathbf{e}_X + L \mathbf{e}_Z \right)\times \mathbf{F}_C - \mathbf{M}^i(x) = \mathbf{0} \implies \mathbf{M}^i(x) =  \left((L-x)\mathbf{e}_X + L \mathbf{e}_Z \right)\times (-P \mathbf{e}_Y) \\
\implies \mathbf{M}^i(x) =  P\left((x-L)\mathbf{e}_Z + L \mathbf{e}_X \right) \implies m_x=\mathbf{M}^i(x)\cdot \mathbf{e}_X= \mathbf{M}^i(x)\cdot \mathbf{e}_x(x) = PL \quad m_z(x)=\mathbf{M}^i(x)\cdot \mathbf{e}_z= \mathbf{M}^i(x)\cdot \mathbf{e}_z = P(x-L)
$$

- Para uma seção $x$ na barra $BC$:
$$
\mathbf{F}_C - \mathbf{F}^i(x) = \mathbf{0} \implies \mathbf{F}^i(x) =  - P \mathbf{e}_Y \implies V_y = - P \mathbf{e}_Y\cdot\mathbf{e}_y = - P \mathbf{e}_Y\cdot\mathbf{e}_Y = - P
$$

$$
\left((L-x)\mathbf{e}_Z \right)\times \mathbf{F}_C - \mathbf{M}^i(x) = \mathbf{0} \implies \mathbf{M}^i(x) =  \left((L-x)\mathbf{e}_Z  \right)\times (-P \mathbf{e}_Y) \\
\implies \mathbf{M}^i(x) =  P\left((2L-x)\mathbf{e}_X  \right) \implies m_x(x)=\mathbf{M}^i(x)\cdot \mathbf{e}_x= \mathbf{M}^i(x)\cdot \mathbf{e}_Z(x) = 0 \quad m_z(x)=\mathbf{M}^i(x)\cdot \mathbf{e}_z= \mathbf{M}^i(x)\cdot (-\mathbf{e}_X) = , \quad  m_z(x)=\mathbf{M}^i(x)\cdot \mathbf{e}_z= \mathbf{M}^i(x)\cdot (-\mathbf{e}_Z) = -P(2L-x)
$$

ii) Não seção de engaste:
$$
\sigma_x = -\frac{m_z}{I_z} y \quad \tau_{xy}= - \frac{m_x}{J} z \quad \tau_{xz}=  \frac{m_x}{J} y
$$

iii)

iv) Deflexão vertical  da extermidade livre:
$$
\delta = \frac{\partial U}{\partial P}
$$
onde
$$
U = \int_0^L \left(\frac{1}{2E I_z} m_z^2  + \frac{1}{2G J} m_x^2\right)\, dx + \int_L^{2L} \left(\frac{1}{2E I_z} m_z^2 \right)\, dx
$$

1) Uma   placa circular (ou cilindro curto)  de raio  externo  $a$ e espessura $h$  possui um orifício central de  raio $c$.  A placa é  submetida a uma pressão externa $p$ conforme ilustrado na figura; não existe força axial.  Neste caso,  o estado de tensões é plano e  definido pelas tensões    radial $\sigma_r(r)=A - B/{r^2}$ e circunferencial $\sigma_\theta(r) = A + {B}/{r^2}$,  onde $A$ e $B$  são constantes a serem determinadas. Para o caso em questão: i) Determine $A$, $B$  e  o diagrama de Mohr para o estado de tensões em $r=c$;  ii)  Determine as  variações da espessura e dos raios interno e  externo da placa  como resultado da aplicação da pressão;  iii)  Obtenha e plote a tensão equivalente de  Tresca em função da coordenada radial $r$. Indique os pontos críticos de acordo com o critério de Tresca; iv)   Calcule a tensão  equivalente de Tresca máxima no limite quando $c/a\rightarrow 0$; v) Calcule a razão entre esta tensão equivalente máxima e a que existiria caso a placa fosse sólida ($c=0$). Discuta este resultado. Dados: $p$, $a$, $c$, $h$, $E$, $\nu$.

## Solução:


i) Determinação de $A$ e $B$:

Usando as condições de contorno $\sigma_r(c)=0$ e $\sigma_r(a)=-p$ chegamos ao sistema
$$
\left\{
\begin{array}{lcr}
 A  - \frac{B}{c^2} = 0 \cr
A  - \frac{B}{a^2} = -p
\end{array}
\right. \iff
\left[
\begin{array}{lcr}
 1 & -1/c^2   \cr
 1 & -1/a^2
\end{array}
\right] \left[
\begin{array}{lcr}
 A   \cr
 B
\end{array}
\right] =\left[
\begin{array}{lcr}
 0   \cr
 -p
\end{array}
\right]
$$
Logo:
$$
\left\{
\begin{array}{lcr}
 A  = - \frac{a^2}{a^2-c^2} p  \cr
B = - \frac{a^2 c^2}{a^2-c^2} p
\end{array}
\right.
$$


ii) Variação da espessura $h$:
$$
\frac{\triangle h}{h} = \epsilon_z = \frac{1}{E}\left(\sigma_z - \nu(\sigma_r + \sigma_\theta)\right)= - \frac{\nu}{E} 2 A = \frac{\nu}{E} \frac{2 a^2}{a^2-c^2} p;
$$
Variação do raio interno:
$$
\frac{\triangle c}{c} = \epsilon_\theta(c) = \frac{1}{E}\left(\sigma_\theta(c) - \nu(\sigma_r(c) + \sigma_z)\right)= \frac{1}{E}\left( A + \frac{B}{c^2} \right) ;
$$
Variação do raio externo:
$$
\frac{\triangle c}{a} = \epsilon_\theta(a) = \frac{1}{E}\left(\sigma_\theta(a) - \nu(\sigma_r(a) + \sigma_z)\right)= \frac{1}{E}\left( A + \frac{B}{a^2} +\nu p \right) ;
$$

iii) Tensão equivalente de Tresca:
$$
\sigma_{eq}(r) = \sigma_1(r) - \sigma_3(r).
$$
No presente caso, podemos observar que
$$
\sigma_1(r)=\sigma_z = 0 \ge \sigma_2(r) = \sigma_r(r) \ge \sigma_3(r)=\sigma_\theta(r).
$$
Lego:
$$
\sigma_{eq}(r) = -\sigma_\theta(r) = \frac{ p a^2 }{a^2 - c^2}\left( 1 + \frac{c^2}{r^2}\right).
$$
Como o maior valor de $\sigma_{eq}$  é alcançado em $r=c$, os pontos críticos estão localizados na superfície interna da placa e a tensão equivalente máxima é:
$$
\max \sigma_{eq} = \sigma_{eq}(c) = \frac{ 2 p a^2 }{a^2 - c^2}.
$$

iv) No limite quando $c/a \rightarrow 0$,
$$
\max \sigma_{eq}=\frac{ 2 p a^2  }{a^2 - c^2} = \frac{ 2 p  }{1 - (c/a)^2} \rightarrow  2p.
$$

v) Quando $c=0$, $B=0$ e portanto $\sigma_r=\sigma_\theta=-p$. Neste caso,
$$
\sigma_1=\sigma_z=0, \quad \sigma_2=\sigma_3=-0 \implies \sigma_{eq}^{c=0}=p
$$
